# Layer 2. Cohort Retention & Customer Value Dynamics

**Бизнес-контекст:**

В условиях сезонного снижения покупательской активности аптечная сеть стремится глубже понять поведенческую динамику клиентов и эффективность удержания на различных этапах жизненного цикла.

Клиентская база неоднородна: пользователи существенно различаются по скорости оттока, устойчивости повторных покупок и динамике потребления после первой транзакции. Это требует перехода от статического анализа продаж к когортному и lifecycle-аналитическому подходу.

**Основная задача раздела:**

Понять, каким образом формируется долгосрочная ценность клиента после первой покупки и какие механизмы удержания и монетизации определяют экономику клиентской базы.

В отличие от традиционного retention-анализа, который фокусируется исключительно на сохранении пользователей, данный раздел рассматривает удержание как один из компонентов customer value creation.

Для этого анализируются:  
- скорость снижения активности клиентов после первой покупки;  
- динамика повторных покупок;  
- изменение выручки внутри удержанной базы;  
- различия между пользовательским и денежным удержанием;  
- эволюция ценности клиента на протяжении жизненного цикла;  
- устойчивость выявленных закономерностей между когортами.  

**Исследовательские вопросы:**  

1. Как быстро снижается активность пользователей после первой покупки?  
2. Формируется ли стабильное повторное покупательское поведение (habitual behavior)?  
3. Как различаются когортные группы по уровню и динамике retention?  
4. Насколько быстро меняется структура выручки среди удержанных пользователей?  
5. В каких точках жизненного цикла происходит максимальная потеря клиентской базы?  
6. Насколько устойчивы различия в поведении между когортами?
7. Что в большей степени определяет долгосрочную ценность клиентской базы:  
    - сохранение пользователей;  
    - или рост монетизации удержанных клиентов?  

**Гипотезы:**    

1. Основной отток клиентов происходит в первые 1–2 месяца после первой покупки.  
2. Часть пользователей после раннего оттока формирует стабильное повторное покупательское поведение.  
3. Динамика revenue retention отличается от user retention и может указывать на рост ценности удержанных клиентов.  
4. Поведение когорт становится более вариативным на поздних этапах жизненного цикла из-за различий в типах покупателей (разовые vs регулярные, хронические пользователи).  
5. Стабилизация retention decay может свидетельствовать о формировании устойчивой пользовательской базы.
6. Revenue Retention будет снижаться медленнее, чем User Retention, поскольку наиболее ценные клиенты остаются активными дольше и генерируют непропорционально большую долю выручки.
7. По мере созревания когорт рост ценности удержанных клиентов частично компенсирует снижение размера клиентской базы.

**Методология:**

Анализ реализован в формате cohort-based lifecycle analysis.

Для каждого пользователя формируются:
- *cohort_month* — месяц первой покупки,  
- *cohort_index* — количество месяцев с момента первой покупки.  

Пользователи агрегируются по когортам и lifecycle-периодам для расчета следующих метрик:
- *User Retention* — доля активных пользователей относительно размера когорты;  
- *Revenue Retention* — динамика выручки относительно первого месяца;  
- *Revenue per Retained User (RPRU)* — средняя выручка на удержанного пользователя;  
- *Retention Decay* — динамика изменения retention во времени (relative и absolute);  
- *Cohort Variability* — вариативность поведения между когортами.  

Дополнительно проводится корректировка cohort maturity за счет исключения незрелых когорт, а также анализ статистической устойчивости метрик с использованием стандартного отклонения и confidence bands.

**В рамках анализа реализованы**

1. Cohort Retention Analysis  
Исследование динамики удержания клиентов по когортам первой покупки.

2. Retention Decay Analysis   
Оценка скорости снижения retention и выявление точек стабилизации пользовательского поведения.

3. Revenue Retention Analysis  
Исследование сохранения выручки внутри когорт независимо от сокращения клиентской базы.

4. Customer Value Evolution Analysis  
Анализ изменения Revenue per Retained User (RPRU) как индикатора роста ценности удержанных клиентов.

5. Retention vs Monetization Analysis  
Сравнение вклада удержания клиентов и роста монетизации в формирование выручки когорт.

6. Cohort LTV Dynamics  
Оценка накопления долгосрочной ценности клиента на протяжении жизненного цикла.

7. Cohort Variability Analysis  
Проверка устойчивости выявленных закономерностей между различными когортами.

**Бизнес-интерпретация:**

Когортный анализ позволяет перейти от оценки текущего состояния клиентской базы к пониманию механики формирования долгосрочной выручки.

Вместо ответа на вопрос:

анализ отвечает на более важные бизнес-вопросы:  
- когда происходит основная потеря клиентов;  
- какие пользователи продолжают приносить выручку спустя месяцы после первой покупки;  
- насколько эффективно компания удерживает экономически ценных клиентов;  
- каким образом формируется Customer Lifetime Value;  
- какие механизмы обеспечивают долгосрочную устойчивость клиентской базы.  

Особую ценность представляет совместный анализ User Retention и Revenue Retention, позволяющий различать количественное удержание клиентов и сохранение их экономической ценности.

**Практический результат:**  

Результатом анализа становится формирование целостной модели жизненного цикла клиента, описывающей взаимосвязь между удержанием пользователей, динамикой выручки и накоплением клиентской ценности.

Полученные выводы позволяют:  
- выявлять критические периоды жизненного цикла;

Результаты данного раздела формируют аналитическую основу для последующего построения моделей прогнозирования оттока и систем customer health monitoring.

Для повышения надежности выводов анализ дополнительно учитывает зрелость когорт и статистическую устойчивость наблюдаемых закономерностей.

## Cohort Retention Framework

На данном этапе переходим к cohort-based retention analysis для исследования долгосрочной динамики пользовательского поведения после первой покупки. В отличие от агрегированных метрик продаж, когортный подход позволяет анализировать lifecycle пользователей во времени и оценивать, насколько устойчиво формируется повторное покупательское поведение внутри различных acquisition cohorts.

Цель анализа — определить скорость снижения пользовательской активности после первой транзакции, выявить наличие retention stabilization effects и сравнить динамику удержания и монетизации между когортами. Дополнительно проверяется гипотеза о том, что revenue retention может снижаться медленнее user retention за счет роста customer value среди удержанных пользователей.

Для этого пользователи группируются по месяцу первой покупки (cohort_month), после чего для каждой когорты рассчитывается lifecycle horizon — количество месяцев с момента привлечения (cohort_index). На основе когортной структуры будут построены две ключевые retention-модели: user retention — доля активных пользователей относительно исходного размера когорты, и revenue retention — динамика выручки относительно revenue acquisition period. Такой подход позволяет одновременно оценить как сохранение поведения, так и монетизационную устойчивость клиентской базы.

Особое внимание уделяется корректности lifecycle-интерпретации: из анализа исключается незавершенный последний месяц наблюдения, поскольку неполные периоды могут искажать retention curves и создавать ложный сигнал ускоренного churn. Когортная модель также позволяет отделить краткосрочное transactional-поведение от формирования устойчивого паттерна повторных покупок.

Если гипотеза подтвердится и revenue retention окажется устойчивее user retention, это будет указывать на концентрацию monetization внутри retained customer base и постепенный рост ценности удержанных пользователей во времени. Подобная динамика особенно важна для оценки долгосрочной устойчивости retention-driven growth модели.

Далее сформируем cohort-таблицы и рассчитаем основные retention-метрики для последующего анализа lifecycle-динамики и развитие клиентской ценности.

```python
# =====================================================
# IMPORTS
# =====================================================

import pandas as pd
import numpy as np
import seaborn as sns
import plotly.graph_objects as go
import matplotlib.pyplot as plt

# =====================================================
# LOAD DATA
# =====================================================

df = pd.read_csv('no_outliers_freq.csv')

df['datetime'] = pd.to_datetime(df['datetime'])

# =====================================================
# PREPARE MONTH COLUMNS
# =====================================================

df['purchase_month'] = (
    df['datetime']
    .dt.to_period('M')
    .dt.to_timestamp()
)

# =====================================================
# REMOVE INCOMPLETE MONTH (JUNE)
# =====================================================

max_month = df['purchase_month'].max()

df = df[df['purchase_month'] < max_month]

# =====================================================
# ACQUISITION MONTH
# =====================================================

first_purchase = (
    df.groupby('card')['purchase_month']
    .min()
    .reset_index()
)

first_purchase.columns = [
    'card',
    'cohort_month'
]

# merge cohort month
df = df.merge(
    first_purchase,
    on='card',
    how='left'
)

# =====================================================
# COHORT INDEX
# =====================================================

df['cohort_index'] = (
    (
        df['purchase_month'].dt.year
        - df['cohort_month'].dt.year
    ) * 12
    +
    (
        df['purchase_month'].dt.month
        - df['cohort_month'].dt.month
    )
)

# =====================================================
# USER RETENTION
# =====================================================

# unique users by cohort + month
cohort_users = (
    df.groupby([
        'cohort_month',
        'cohort_index'
    ])['card']
    .nunique()
    .reset_index()
)

# cohort sizes
cohort_sizes = (
    cohort_users[
        cohort_users['cohort_index'] == 0
    ][[
        'cohort_month',
        'card'
    ]]
    .rename(columns={
        'card': 'cohort_size'
    })
)

# merge sizes
cohort_users = cohort_users.merge(
    cohort_sizes,
    on='cohort_month'
)

# retention %
cohort_users['retention_rate'] = (
    cohort_users['card']
    / cohort_users['cohort_size']
)

# pivot table
user_retention = cohort_users.pivot_table(
    index='cohort_month',
    columns='cohort_index',
    values='retention_rate'
)

# =====================================================
# REVENUE RETENTION
# =====================================================

# revenue by cohort + month
cohort_revenue = (
    df.groupby([
        'cohort_month',
        'cohort_index'
    ])['summ_with_disc']
    .sum()
    .reset_index()
)

# initial cohort revenue
initial_revenue = (
    cohort_revenue[
        cohort_revenue['cohort_index'] == 0
    ][[
        'cohort_month',
        'summ_with_disc'
    ]]
    .rename(columns={
        'summ_with_disc': 'initial_revenue'
    })
)

# merge
cohort_revenue = cohort_revenue.merge(
    initial_revenue,
    on='cohort_month'
)

# revenue retention %
cohort_revenue['revenue_retention'] = (
    cohort_revenue['summ_with_disc']
    / cohort_revenue['initial_revenue']
)

# pivot
revenue_retention = cohort_revenue.pivot_table(
    index='cohort_month',
    columns='cohort_index',
    values='revenue_retention'
)
```

## User Retention Dynamics Across Cohorts

На данном этапе исследуется динамика user retention между acquisition-когортами, чтобы оценить устойчивость повторного покупательского поведения после первой покупки. Анализ retention-кривых позволяет перейти от агрегированных метрик к пониманию lifecycle-паттернов отдельных когорт.

Цель анализа — определить, насколько быстро снижается пользовательская активность после acquisition, существуют ли различия в скорости оттока между когортами и формируется ли стабильное ядро удержанных пользователей на более поздних этапах жизненного цикла. Дополнительно проверяется гипотеза о концентрации основного churn в первые месяцы после первой покупки.

Для анализа используется cohort-based retention approach: для каждой acquisition-когорты рассчитывается доля активных пользователей в последующие lifecycle-периоды относительно исходного размера когорты. Визуализация retention curves позволяет сравнить скорость decay, устойчивость удержания и степень неоднородности пользовательского поведения между когортами.

Сходная форма retention-кривых может указывать на стабильность моделей поведения внутри продукта, тогда как значительные расхождения между когортами способны сигнализировать об изменениях качества acquisition, сезонности или различиях в пользовательских сценариях потребления.

Далее визуализируется retention curves для каждой acquisition-когорты и сравнивается динамика удержания пользователей во времени.

```python
# =====================================================
# USER RETENTION CURVES
# =====================================================

plt.figure(figsize=(10,6))

for cohort in user_retention.index:

    plt.plot(
        user_retention.columns,
        user_retention.loc[cohort],
        marker='o',
        label=str(cohort.date())
    )

plt.title('User Retention Curves', fontsize=14)
plt.xlabel('Months Since Acquisition')
plt.ylabel('User Retention Rate')

plt.legend(
    title='Cohort',
    bbox_to_anchor=(1.05,1),
    loc='upper left'
)

plt.grid(True)
plt.tight_layout()
plt.show()
```

## Revenue Retention Dynamics Across Cohorts

После оценки user retention необходимо проанализировать динамику revenue retention, чтобы понять, как изменяется способность когорт генерировать выручку на разных этапах жизненного цикла. User retention отражает сохранение аудитории, однако не показывает изменение качества монетизации удержанных пользователей.

Цель анализа — определить, насколько быстро снижается revenue contribution после acquisition, отличаются ли revenue retention patterns между когортами и компенсируется ли пользовательский churn ростом spending среди retained users. Дополнительно проверяется гипотеза о расхождении между user retention и revenue retention как индикаторе изменения customer value во времени.

Для анализа используется cohort-based revenue retention подход: для каждой acquisition-когорты рассчитывается доля выручки относительно revenue первого месяца жизни когорты. Визуализация retention curves позволяет сравнить скорость monetization decay, устойчивость revenue generation и вариативность монетизационного поведения между когортами.

Если revenue retention снижается медленнее user retention, это может указывать на концентрацию выручки среди более ценных и вовлеченных пользователей. Напротив, синхронное падение обеих метрик способно сигнализировать о слабом формировании долгосрочной customer value.

Далее визуализируется revenue retention curves для сравнения монетизационной динамики между acquisition-когортами.

```python
# =====================================================
# REVENUE RETENTION CURVES
# =====================================================

plt.figure(figsize=(10,6))

for cohort in revenue_retention.index:

    plt.plot(
        revenue_retention.columns,
        revenue_retention.loc[cohort],
        marker='o',
        label=str(cohort.date())
    )

plt.title('Revenue Retention Curves', fontsize=14)
plt.xlabel('Months Since Acquisition')
plt.ylabel('Revenue Retention Rate')

plt.legend(
    title='Cohort',
    bbox_to_anchor=(1.05,1),
    loc='upper left'
)

plt.grid(True)
plt.tight_layout()
plt.show()
```

## Cohort Retention Matrices

После анализа retention curves необходимо детализировать поведение когорт на уровне отдельных lifecycle-периодов. Линейные графики позволяют оценить общую динамику retention, однако ограниченно показывают локальные изменения между когортами и этапами жизненного цикла.

Цель анализа — выявить точки максимального churn, определить периоды стабилизации retention и оценить неоднородность пользовательского и revenue behavior между acquisition-когортами. Дополнительно проверяется, сохраняется ли различие между user retention и revenue retention на поздних стадиях lifecycle.

Для анализа используются cohort heatmaps, где каждая ячейка отражает retention-уровень конкретной когорты в соответствующий lifecycle-период. Такой подход упрощает сравнение качества когорт, скорости retention decay и устойчивости модели монетизации между когортами.

Высокая концентрация снижения retention в ранних периодах может указывать на слабую post-acquisition activation или низкую привычку к повторным покупкам. Более стабильные значения revenue retention по сравнению с user retention способны свидетельствовать о росте customer value среди удержанных пользователей.

Далее визуализируется cohort heatmaps для user retention и revenue retention с целью сравнения lifecycle dynamics между acquisition-когортами.

```python
# =====================================================
# PLOT
# =====================================================

user_retention.index = user_retention.index.strftime('%Y-%m')
revenue_retention.index = revenue_retention.index.strftime('%Y-%m')

plt.figure(figsize=(14, 6))
plt.title('User Retention Table', fontsize=14)

sns.heatmap(
    data=user_retention,
    annot=True,
    fmt='.0%',
    vmin=0.0,
    vmax=0.5,
    linewidths=.4,
    cmap='Blues'
)

plt.yticks(rotation=0)  # optional cleaner look
plt.xlabel('Months Since Acquisition')
plt.ylabel('Acquisition Cohort')
plt.show()

plt.figure(figsize=(14, 6))
plt.title('Revenue Retention Table', fontsize=14)

sns.heatmap(
    data=revenue_retention,
    annot=True,
    fmt='.0%',
    linewidths=.4,
    cmap='BuGn'
)

plt.yticks(rotation=0)
plt.xlabel('Months Since Acquisition')
plt.ylabel('Acquisition Cohort')

plt.tight_layout()
plt.show()
```

**Key Findings**

Когортный анализ показал устойчивый retention decay во всех acquisition-когортах. User Retention после первого месяца снижается с 100% до ~19–34%, после чего темпы оттока замедляются и на горизонте 6–10 месяцев стабилизируются в диапазоне 15–30%. Revenue Retention демонстрирует схожую динамику, однако в большинстве когорт остается выше User Retention и на поздних этапах жизненного цикла удерживается на уровне 20–45% от выручки первого месяца.

Основная потеря аудитории происходит в течение первого месяца после первой покупки, что указывает на слабую конверсию новых клиентов в регулярных покупателей. При этом более медленное снижение Revenue Retention относительно User Retention свидетельствует о концентрации выручки среди удержанных пользователей: по мере оттока аудитории в продукте остается более ценное ядро клиентов, генерирующее непропорционально большую долю дохода. Для бизнеса это означает, что долгосрочная монетизация определяется не масштабом привлечения как таковым, а способностью переводить новых покупателей в сегмент повторных клиентов.

Возможной причиной высокого раннего churn может быть завершение единичного покупательского сценария без формирования устойчивой потребительской привычки. Также наблюдаемая концентрация выручки предположительно связана с наличием сегмента лояльных клиентов, отличающихся более высокой частотой покупок и средним чеком.

Ключевое ограничение growth-модели находится не в монетизации удержанных пользователей, а в раннем удержании после первой покупки. Потенциал роста LTV, вероятно, в большей степени зависит от улучшения post-acquisition retention, чем от дополнительной оптимизации среднего чека существующей лояльной базы.

Приоритетными направлениями дальнейшего анализа и оптимизации являются:   
1. Исследовать причины churn между Month 0 и Month 1 как основной источник потери клиентской базы.  
2. Сегментировать удержанных пользователей для выявления behavioral patterns, связанных с высокой долгосрочной ценностью.  
3. Оценить эффективность программ повторной активации в первые 30–60 дней после первой покупки.  
4. Сравнить retention и revenue retention между acquisition-периодами для выявления факторов, влияющих на качество новых когорт.  

Рост раннего retention способен увеличить размер долгосрочного активного ядра клиентов, что приведет к росту cumulative LTV, улучшению unit-экономики и повышению эффективности инвестиций в привлечение новых покупателей.

## Retention Stability & Lifecycle Decay Analysis

После анализа отдельных когорт необходимо оценить общую структуру lifecycle-динамики и устойчивость retention-поведения на уровне всей клиентской базы. На данном этапе исследуется скорость снижения удержания, степень различий между когортами и наличие признаков стабилизации пользовательского поведения во времени.

Цель анализа — определить, в каких lifecycle-периодах происходит максимальный отток пользователей и выручки, насколько быстро замедляется retention decay, а также насколько устойчиво поведение когорт между собой. Дополнительно проверяется гипотеза о том, что после раннего периода churn retention переходит в более стабильную фазу, отражающую формирование ядра регулярных клиентов.

Для повышения корректности интерпретации из анализа исключаются незрелые когорты с недостаточным lifecycle-горизонтом. Далее рассчитываются средние user retention и revenue retention, month-over-month decay, абсолютные retention drops, а также variability между когортами через standard deviation и confidence bands. Такой подход позволяет одновременно оценить среднюю траекторию удержания и степень неоднородности поведения клиентов.

Ожидается, что наиболее резкое снижение retention будет наблюдаться в первые месяцы после первой покупки, тогда как дальнейшая динамика станет более плавной. Снижение variability на поздних этапах lifecycle может указывать на формирование более устойчивого сегмента retained клиентов, тогда как рост разброса между когортами — на усиление различий между типами покупателей и сценариями потребления.

```python
# =====================================================
# FILTER IMMATURE COHORTS
# =====================================================

mature_user_retention = user_retention.iloc[:-2]
mature_revenue_retention = revenue_retention.iloc[:-2]
mature_rpru_table = rpru_table.iloc[:-2]

avg_user_retention = mature_user_retention.mean()
avg_revenue_retention = mature_revenue_retention.mean()

# =====================================================
# RETENTION DECAY ANALYSIS
# =====================================================

retention_decay = (
    avg_user_retention
    .pct_change()
)

revenue_decay = (
    avg_revenue_retention
    .pct_change()
)

retention_drop = (
    avg_user_retention
    .diff()
)

revenue_drop = (
    avg_revenue_retention
    .diff()
)

# =====================================================
# NUMBER OF COHORTS CONTRIBUTING
# =====================================================

cohort_counts = (
    mature_user_retention
    .count()
)

# =====================================================
# RETENTION VARIABILITY
# =====================================================

retention_std = (
    mature_user_retention
    .std()
)

revenue_retention_std = (
    mature_revenue_retention
    .std()
)

retention_upper = (
    avg_user_retention
    + retention_std
)

retention_lower = (
    avg_user_retention
    - retention_std
)

revenue_upper = (
    avg_revenue_retention
    + revenue_retention_std
)

revenue_lower = (
    avg_revenue_retention
    - revenue_retention_std
)

# =====================================================
# AVERAGE RETENTION WITH VARIABILITY
# =====================================================

fig, ax1 = plt.subplots(figsize=(10,6))

ax1.plot(
    avg_user_retention.index,
    avg_user_retention.values,
    marker='o',
    linewidth=2,
    color='blue',
    label='User Retention'
)

ax1.plot(
    avg_revenue_retention.index,
    avg_revenue_retention.values,
    marker='o',
    linewidth=2,
    color='green',
    label='Revenue Retention'
)

# confidence bands
ax1.fill_between(
    avg_user_retention.index,
    retention_lower,
    retention_upper,
    color='blue',
    alpha=0.15
)

ax1.fill_between(
    avg_revenue_retention.index,
    revenue_lower,
    revenue_upper,
    color='green',
    alpha=0.15
)

ax1.set_xlabel('Months Since Acquisition')

ax1.set_ylabel('Retention Rate')

ax1.grid(True)

ax2 = ax1.twinx()

ax2.bar(
    cohort_counts.index,
    cohort_counts.values,
    alpha=0.15,
    color='gray',
    label='Cohorts Included'
)

ax2.set_ylabel('Number of Cohorts')

lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()

ax1.legend(
    lines1 + lines2,
    labels1 + labels2,
    loc='upper right'
)

plt.title('Average Retention with Cohort Variability', fontsize=14)
plt.tight_layout()
plt.show()

# =====================================================
# RETENTION DECAY CURVES
# =====================================================

plt.figure(figsize=(10,6))

plt.plot(
    retention_decay.index,
    retention_decay.values,
    marker='o',
    linewidth=2,
    label='User Retention Decay'
)

plt.plot(
    revenue_decay.index,
    revenue_decay.values,
    marker='o',
    linewidth=2,
    label='Revenue Retention Decay'
)

plt.axhline(
    y=0,
    linestyle='--',
    color='gray'
)

plt.title('Month-over-Month Retention Decay', fontsize=14)

plt.xlabel('Months Since Acquisition')

plt.ylabel('Relative Change')

plt.legend()

plt.grid(True)
plt.tight_layout()
plt.show()

# =====================================================
# RETENTION DROP CURVES
# =====================================================

plt.figure(figsize=(10,6))

plt.plot(
    retention_drop.index,
    retention_drop.values,
    marker='o',
    linewidth=2,
    label='User Retention Drop'
)

plt.plot(
    revenue_drop.index,
    revenue_drop.values,
    marker='o',
    linewidth=2,
    label='Revenue Retention Drop'
)

plt.axhline(
    y=0,
    linestyle='--',
    color='gray'
)

plt.title('Absolute Retention Drop', fontsize=14)

plt.xlabel('Months Since Acquisition')

plt.ylabel('Percentage Point Change')

plt.legend()

plt.grid(True)
plt.tight_layout()
plt.show()
```

**Key Findings**

Анализ средних retention-кривых показал крайне неравномерный характер оттока на протяжении жизненного цикла. Основное снижение как user retention, так и revenue retention происходит между первым и вторым месяцем после первой покупки: средняя потеря составляет около 74 п.п. по пользователям и 71 п.п. по выручке. После этого скорость снижения резко замедляется, а изменения между соседними периодами становятся минимальными и в большинстве случаев не превышают нескольких процентных пунктов.

Динамика user retention и revenue retention остаётся близкой на протяжении всего наблюдаемого горизонта. При этом после первых месяцев жизненного цикла обе метрики демонстрируют признаки стабилизации, а абсолютные изменения становятся близкими к нулю, что указывает на формирование относительно устойчивой базы регулярных покупателей.

Результаты показывают, что ключевая проблема продукта находится не в долгосрочном удержании клиентов, а в прохождении раннего post-acquisition периода. Большая часть будущей клиентской базы теряется сразу после первой покупки, тогда как пользователи, успешно преодолевшие этот этап, демонстрируют существенно более стабильное поведение.

Фактически жизненный цикл клиента разделяется на две разные фазы: фазу первичной селекции и фазу удержания лояльной аудитории. Именно первая фаза определяет будущий объём retention, revenue и итогового LTV. После формирования ядра клиентов дальнейшее снижение активности становится значительно менее выраженным и оказывает ограниченное влияние на экономику продукта.

Возможной причиной столь резкого раннего churn может быть отсутствие сформированной покупательской привычки после первой транзакции, недостаточная скорость возврата клиента к повторной покупке либо низкая эффективность механизмов post-purchase engagement. Также эффект может быть связан с высокой долей пользователей, совершающих разовые или ситуативные покупки без намерения переходить в регулярное потребление.

Основной потенциал роста находится в сокращении оттока между первым и вторым месяцем жизни клиента. На текущем этапе продукт выглядит скорее как система с высокой потерей новых покупателей, чем как система со слабым долгосрочным удержанием.

Приоритетными направлениями дальнейшего анализа и оптимизации являются:  
1. Сконцентрировать retention-инициативы на первых 30–60 днях после первой покупки.  
2. Исследовать путь клиентов, не совершающих вторую покупку, и выявить основные точки разрыва в lifecycle.  
3. Разработать механики ускорения второй покупки (персонализированные предложения, триггерные коммуникации, программы возврата).  
4. Отдельно сегментировать retention по категориям товаров и источникам привлечения для поиска когорт с наилучшей конверсией в повторную покупку.  

Даже небольшое снижение churn в интервале между первым и вторым месяцем способно оказать непропорционально сильное влияние на последующие retention-метрики, cumulative revenue и итоговый LTV, поскольку увеличивает размер клиентского сегмента, который переходит в стабильную фазу жизненного цикла.

## Revenue Quality Analysis per Retained User

На данном этапе анализ смещается с оценки самого retention на качество монетизации удержанной аудитории. Даже при снижении количества активных пользователей продукт может сохранять или увеличивать revenue contribution за счет роста spend intensity среди remaining users.

Цель анализа — определить, как меняется средняя выручка на удержанного пользователя (Revenue per Retained User, RPRU) на разных этапах lifecycle, а также проверить, компенсирует ли рост монетизации retained-аудитории естественный retention decay. Дополнительно анализ позволяет выявить различия в purchasing behavior между когортными группами и оценить устойчивость клиентской ценности после первой покупки.

Для анализа объединяются cohort-level данные по retention и revenue, после чего рассчитывается средняя выручка на одного активного пользователя внутри каждой cohort-month комбинации. Когортный подход позволяет отделить эффект сокращения пользовательской базы от изменений в spend behavior retained users и оценить эволюцию клиентской ценности во времени.

Рост RPRU на поздних lifecycle-этапах может указывать на формирование более качественного и вовлеченного customer segment, тогда как снижение метрики способно сигнализировать об ослаблении purchase intensity даже среди удержанных пользователей. Для интерпретации динамики будет построена cohort heatmap monetization behavior.

Рассчитаем Revenue per Retained User для каждой когорты и визуализируем lifecycle-динамику монетизации.

```python
# =====================================================
# REVENUE PER RETAINED USER
# =====================================================

cohort_metrics = cohort_users.merge(
    cohort_revenue,
    on=['cohort_month', 'cohort_index']
)

cohort_metrics['revenue_per_user'] = (
    cohort_metrics['summ_with_disc']
    /
    cohort_metrics['card']
)

rpru_table = cohort_metrics.pivot_table(
    index='cohort_month',
    columns='cohort_index',
    values='revenue_per_user'
)

# =====================================================
# PLOT
# =====================================================

rpru_table.index = rpru_table.index.strftime('%Y-%m')

plt.figure(figsize=(14,6))

heatmap = sns.heatmap(
    rpru_table,
    annot=True,
    fmt='.0f',
    cmap='Greens',
    linewidths=.4,
    annot_kws={'fontsize': 10}
)

for text in heatmap.texts:
    val = float(text.get_text())
    text.set_text(f'{val/1000:.1f}k')
    
plt.title('Revenue per Retained User', fontsize=14)
plt.xlabel('Months Since Acquisition')
plt.ylabel('Acquisition Cohort')
plt.tight_layout()
plt.show()
```

**Key Findings**

Анализ Revenue per Retained User показал, что после первого месяца жизненного цикла средняя выручка на удержанного пользователя не снижается вместе с retention, а остается стабильной или увеличивается. Для большинства когорт RPRU находится в диапазоне 1.2–2.1 тыс. денежных единиц на пользователя и зачастую превышает уровень первого месяца. Наиболее зрелые когорты демонстрируют сохранение или рост интенсивности расходов даже на поздних lifecycle-этапах, несмотря на существенное сокращение пользовательской базы.

Полученные результаты показывают, что retention decay сопровождается естественной селекцией аудитории: продукт теряет значительную часть пользователей, но сохраняет наиболее ценных клиентов с более высоким уровнем потребления. Снижение количества активных пользователей частично компенсируется ростом customer value среди retained-клиентов, что объясняет более устойчивую динамику Revenue Retention по сравнению с User Retention. Для бизнеса это означает наличие сформированного сегмента лояльных клиентов, который генерирует непропорционально высокую долю выручки и является основным драйвером долгосрочного LTV.

Возможной причиной роста RPRU может быть концентрация в поздних lifecycle-периодах пользователей с более высокой частотой покупок, большим средним чеком или более широким ассортиментом потребления. Также наблюдаемая динамика предположительно связана с тем, что низкововлеченные клиенты покидают продукт раньше, тогда как наиболее ценные пользователи сохраняют регулярную покупательскую активность.

Монетизация удержанной аудитории выглядит устойчивой и не является основным ограничением роста. Ключевой резерв развития находится в увеличении доли пользователей, которые достигают стадии долгосрочного удержания и переходят в сегмент высокоценных клиентов.

Приоритетными направлениями дальнейшего анализа и оптимизации являются:  

1. Выделить и профилировать сегмент пользователей с высоким RPRU для выявления факторов долгосрочной ценности.  
2. Исследовать behavioral differences между churned и retained пользователями в первые месяцы после acquisition.  
3. Оценить возможность ускорения перехода новых клиентов в высокоценный сегмент через персонализированные предложения и программы лояльности.  
4. Дополнительно проанализировать вклад frequency и average basket в рост RPRU.  

Увеличение доли пользователей, достигающих стадии высокой монетизации, позволит повысить cumulative LTV без пропорционального роста затрат на привлечение, улучшить unit-экономику и повысить эффективность существующих acquisition-инвестиций.

## Retention vs Monetization Dynamics

После анализа retention stability и cohort-level monetization важно оценить взаимосвязь между удержанием пользователей и качеством монетизации retained-аудитории. На практике снижение активной базы не всегда приводит к пропорциональному снижению customer value: часть пользователей может демонстрировать более высокий уровень интенсивности расходов по мере прохождения lifecycle.

Цель анализа — определить, компенсирует ли рост Revenue per Retained User естественный retention decay, а также оценить, формируется ли среди удержанных пользователей более ценный long-term customer segment. Дополнительно анализ позволяет проверить гипотезу о расхождении между behavioral retention и monetization dynamics.

Для анализа используются усредненные cohort-level метрики mature cohorts: User Retention отражает динамику удержания аудитории, а Average Revenue per Retained User — изменение среднего revenue contribution активных пользователей на разных lifecycle-этапах. Сопоставление этих кривых позволяет оценить баланс между сокращением пользовательской базы и изменением качества монетизации.

Если снижение retention сопровождается ростом RPRU, это может указывать на естественную селекцию более вовлеченных и ценных пользователей. Напротив, одновременное снижение обеих метрик способно сигнализировать об ослаблении клиентской вовлеченности и деградации потенциала долгосрочной монетизации.

Сравним динамику retention и монетизацию retained users на протяжении customer lifecycle.

Таким образом, снижение численности активной аудитории сопровождается постепенным ростом customer value среди пользователей, остающихся в продукте.

Результаты подтверждают наличие механизма естественной селекции клиентской базы. Основной churn приходится на пользователей с более низкой вовлечённостью и меньшим вкладом в выручку, тогда как retained-аудитория постепенно концентрирует всё большую долю customer value. Фактически продукт теряет количество пользователей быстрее, чем качество монетизации.

Для бизнеса это означает наличие устойчивого сегмента лояльных клиентов, который формирует непропорционально высокий вклад в долгосрочную выручку. Однако рост RPRU не компенсирует масштаб первоначального оттока: увеличение монетизации на одного retained user составляет порядка 15–40%, тогда как клиентская база сокращается почти на 75% уже после первого месяца. Следовательно, главным ограничителем роста LTV остаётся retention, а не эффективность монетизации.

Возможной причиной наблюдаемой динамики является концентрация повторных покупок среди наиболее вовлечённых клиентов, которые продолжают увеличивать частоту покупок, средний чек или широту потребления ассортимента. Дополнительно эффект может быть связан с тем, что случайные и низкоценные покупатели покидают продукт значительно раньше, чем формируется долгосрочная клиентская база.

Продукт уже демонстрирует способность эффективно монетизировать удержанных пользователей. Ключевая возможность находится не в дополнительном росте интенсивности трат существующих loyal customers, а в увеличении доли пользователей, которые достигают стадий жизненного цикла с повышенным customer value.

Приоритетными направлениями дальнейшего анализа и оптимизации являются:  

1. Снизить churn между первой и второй покупкой как главный драйвер будущего LTV.  
2. Построить профиль high-value retained клиентов и использовать его для оптимизации acquisition и CRM-коммуникаций.  
3. Масштабировать механики, ускоряющие переход новых клиентов в сегмент регулярных покупателей.  
4. Сегментировать RPRU по категориям товаров и частоте покупок для выявления основных драйверов customer value.  

Рост раннего retention позволит увеличить число пользователей, достигающих lifecycle-этапов с повышенным уровнем монетизации. Это приведёт к росту cumulative LTV, улучшению окупаемости привлечения и увеличению долгосрочной выручки без пропорционального роста маркетинговых затрат.

## Cohort Lifetime Value Dynamics

На предыдущих этапах были исследованы механики удержания и изменения монетизации среди активных пользователей. Однако для оценки долгосрочной ценности клиентской базы важно рассматривать не отдельные компоненты lifecycle, а накопленный финансовый эффект, который формируется по мере развития когорты.

Цель анализа — оценить скорость накопления Customer Lifetime Value (LTV) и определить, насколько различаются траектории формирования клиентской ценности между когортами. Дополнительно проверяется гипотеза о том, что существенная часть LTV формируется в первые месяцы после привлечения, после чего темпы прироста постепенно замедляются вследствие снижения retention и покупательской активности.

Для анализа используется когортный подход, в рамках которого рассчитывается накопленная выручка на одного привлечённого пользователя для каждого месяца жизненного цикла. Такой подход позволяет одновременно учитывать влияние удержания, частоты покупок и среднего чека, формируя интегральную оценку экономической ценности клиента во времени.

Опережающий рост отдельных когорт может указывать на более высокое качество привлечённой аудитории или различия в потребительском поведении. Напротив, быстрое насыщение LTV-кривых будет свидетельствовать о раннем исчерпании потенциала монетизации и ограниченном вкладе поздних lifecycle-периодов в формирование общей выручки.

Для оценки динамики накопления клиентской ценности построим когортные кривые Customer Lifetime Value по месяцам жизненного цикла.

```python
# =====================================================
# LOAD DATA
# =====================================================

df = pd.read_csv('no_outliers_freq.csv')

df['datetime'] = pd.to_datetime(df['datetime'])

# =====================================================
# PREPARE MONTH COLUMNS
# =====================================================

df['purchase_month'] = (
    df['datetime']
    .dt.to_period('M')
    .dt.to_timestamp()
)

# =====================================================
# REMOVE INCOMPLETE LAST MONTH
# =====================================================

df = df[df['purchase_month'] < df['purchase_month'].max()]


# =====================================================
# ACQUISITION COHORT
# =====================================================

first_purchase = (
    df.groupby('card')['purchase_month']
    .min()
    .reset_index()
)

first_purchase.columns = [
    'card',
    'cohort_month'
]

df = df.merge(
    first_purchase,
    on='card',
    how='left'
)

# =====================================================
# COHORT INDEX
# =====================================================

df['cohort_index'] = (
    (
        df['purchase_month'].dt.year
        - df['cohort_month'].dt.year
    ) * 12
    +
    (
        df['purchase_month'].dt.month
        - df['cohort_month'].dt.month
    )
)

# =====================================================
# COHORT SIZES
# =====================================================

cohort_sizes = (
    df.groupby('cohort_month')['card']
    .nunique()
    .reset_index()
)

cohort_sizes.columns = [
    'cohort_month',
    'cohort_size'
]

# =====================================================
# MONTHLY REVENUE BY COHORT
# =====================================================

cohort_revenue = (
    df.groupby([
        'cohort_month',
        'cohort_index'
    ])['summ_with_disc']
    .sum()
    .reset_index()
)

# =====================================================
# MERGE COHORT SIZE
# =====================================================

cohort_revenue = cohort_revenue.merge(
    cohort_sizes,
    on='cohort_month',
    how='left'
)

# =====================================================
# REVENUE PER USER
# =====================================================

cohort_revenue['revenue_per_user'] = (
    cohort_revenue['summ_with_disc']
    / cohort_revenue['cohort_size']
)

# =====================================================
# CUMULATIVE LTV
# =====================================================

cohort_revenue['cumulative_ltv'] = (
    cohort_revenue
    .groupby('cohort_month')['revenue_per_user']
    .cumsum()
)

# =====================================================
# PIVOT TABLE
# =====================================================

ltv_table = cohort_revenue.pivot_table(
    index='cohort_month',
    columns='cohort_index',
    values='cumulative_ltv'
)

# =====================================================
# PLOT COHORT LTV CURVES
# =====================================================

plt.figure(figsize=(10,6))

for cohort in ltv_table.index:

    plt.plot(
        ltv_table.columns,
        ltv_table.loc[cohort],
        marker='o',
        label=str(cohort.date())
    )

plt.title('Cohort LTV Curves', fontsize=14)

plt.xlabel('Months Since Acquisition')

plt.ylabel('Cumulative Revenue per User')

plt.legend(
    title='Acquisition Cohort',
    bbox_to_anchor=(1.05,1),
    loc='upper left'
)

plt.grid(True)
plt.tight_layout()
plt.show()
```

Цель анализа — определить, насколько равномерно формируется Customer Lifetime Value между когортами, а также выявить периоды ускоренного или замедленного накопления клиентской ценности. Дополнительно проверяется гипотеза о существовании устойчивых различий в качестве когорт, отражающихся в уровне накопленного LTV на сопоставимых этапах жизненного цикла.

Для анализа используется когортная матрица накопленного LTV, где каждая ячейка отражает суммарную выручку на одного привлечённого пользователя к определённому месяцу жизни когорты. Такой формат позволяет одновременно сравнивать как скорость роста клиентской ценности внутри когорт, так и различия между когортами на одинаковых lifecycle-этапах.

Выраженные различия между строками матрицы могут указывать на неоднородность качества привлечённой аудитории, тогда как замедление изменения значений по мере роста cohort index будет свидетельствовать о постепенном насыщении LTV и снижении вклада поздних периодов жизненного цикла в общую ценность клиента.

Для детального сравнения накопленного Customer Lifetime Value построим когортную LTV-матрицу.

Когортный анализ LTV показал устойчивый рост накопленной ценности клиента во всех когортах на протяжении жизненного цикла. Наиболее зрелые когорты достигают LTV на уровне 6.3–6.8 тыс. денежных единиц к 10–11 месяцу жизни, тогда как основная часть ценности формируется в первые 3–5 месяцев после привлечения. После шестого месяца рост LTV сохраняется, однако темпы накопления постепенно замедляются, что свидетельствует о снижении предельного вклада каждого последующего периода в общую выручку клиента.

Результаты показывают, что продукт способен генерировать долгосрочную клиентскую ценность, однако окно максимального экономического эффекта сосредоточено в раннем lifecycle. Именно первые месяцы после первой покупки определяют значительную часть будущего LTV, поскольку в этот период одновременно происходит и основной churn, и формирование наиболее ценных клиентов. При этом отсутствие резкого насыщения LTV-кривых у зрелых когорт свидетельствует о наличии устойчивого сегмента пользователей, продолжающего генерировать дополнительную выручку даже на поздних этапах жизненного цикла.

Возможной причиной ускоренного накопления LTV в первые месяцы является высокая концентрация повторных покупок сразу после привлечения клиента. Последующее замедление роста предположительно связано с естественным сокращением активной аудитории и переходом продукта в режим монетизации более узкого сегмента лояльных покупателей. Различия между когортами могут быть связаны с изменениями качества acquisition, сезонностью спроса или структурой клиентских сегментов.

Ключевой драйвер роста LTV находится в ранних lifecycle-периодах. Каждый дополнительный процент retention в первые месяцы способен оказывать существенно большее влияние на итоговую ценность клиента, чем аналогичное улучшение на поздних стадиях жизненного цикла.

Приоритетными направлениями дальнейшего анализа и оптимизации являются:  

1. Сфокусировать retention-инициативы на первых 60–90 днях после первой покупки как наиболее значимом периоде формирования LTV.  
2. Выделить характеристики когорт с наиболее высокой скоростью накопления ценности и использовать их для оптимизации acquisition-стратегии.  
3. Исследовать механики, стимулирующие повторные покупки в раннем lifecycle, и масштабировать наиболее эффективные сценарии.  
4. Использовать полученные LTV-кривые как основу для оценки допустимого CAC и долгосрочной unit-экономики.  

Улучшение retention и purchase frequency в первые месяцы жизненного цикла позволит увеличить итоговый LTV, повысить окупаемость привлечения клиентов и создать дополнительный запас для масштабирования acquisition без ухудшения экономики продукта.

## Executive Summary

Когортный анализ показал, что клиентская база характеризуется резким ранним оттоком и последующей стабилизацией поведения удержанных пользователей. Основная потеря клиентов происходит в первые 1–2 месяца после первой покупки: average user retention снижается с 100% до ~26% уже ко второму месяцу. Однако после первичного churn retention curves выходят на плато в диапазоне 20–24%, что указывает на формирование устойчивого ядра регулярных покупателей.

При этом динамика монетизации выглядит существенно сильнее динамики user retention. Несмотря на сокращение активной базы, revenue retention остается выше user retention практически на всем lifecycle horizon, а revenue per retained user со временем растет относительно acquisition month. Это означает, что удержанные пользователи постепенно увеличивают customer value и компенсируют часть потерь клиентской базы за счет более высокой интенсивности трат. В результате долгосрочная стабильность доходов бизнеса определяется не масштабом удержания всей аудитории, а способностью формировать и развивать устойчивый сегмент repeat customers.

### Ключевые инсайты

**1. Основной churn концентрируется в первые месяцы lifecycle**  
User retention падает с 100% до ~26% уже к Month 1–2 практически во всех когортах.
Значительная часть пользователей остается transactional-only customers и не переходит в повторное совершение покупок.
Основной риск потери LTV формируется на этапе early activation и post-first-purchase engagement.

**2. После initial churn формируется стабильное retention plateau**  
После Month 2–3 retention decay существенно замедляется, а absolute retention drop становится близким к нулю.
Оставшиеся пользователи демонстрируют признаки "привычного" покупательского поведения и более устойчивого вовлечения.
Long-term customer base формируется за счет ограниченного, но стабильного сегмента постоянных клиентов.

**3. Revenue retention устойчивее user retention**  
Revenue retention стабильно превышает user retention на поздних lifecycle stages.
Удержанные пользователи со временем становятся более ценными и генерируют непропорционально высокий вклад в revenue.
Потеря low-value пользователей частично компенсируется ростом монетизации retained базы.

**4. Revenue per retained user растет во времени**  
Средний RPRU growth превышает acquisition baseline практически на всех lifecycle стадиях и достигает +43% к поздним периодам.
Среди retained клиентов наблюдается expansion behavior: увеличение частоты, размера корзины или потребности в регулярных покупках.
Бизнес получает органический рост монетизации без необходимости пропорционального расширения клиентской базы.

**5. Variability between cohorts усиливается на поздних стадиях lifecycle**   
Confidence bands становятся шире по мере роста cohort age.
Поведение долгоживущих клиентов становится менее однородным из-за сегментации на эпизодических и highly loyal покупателей.
Монетизация на поздней стадии все больше зависит от качества high-value сегмента клиентов.

**6. Cohort LTV curves демонстрируют накопительный monetization effect**  
Даже при низком retention cumulative LTV продолжает устойчиво расти во всех зрелых когортах.
Retained клиенты продолжают генерировать значимую дополнительную revenue на длинном lifecycle горизонте.
Экономика продукта существенно зависит от долгосрочного удержания повторных покупок, а не только от монетизации первой покупки.

### Интерпретация результатов

Анализ показывает классическую ассиметричную lifecycle структуру, характерную для transactional retail-моделей: большая часть пользователей не формирует устойчивого покупательского поведения после первой покупки, однако небольшое retained-ядро core обеспечивает основную долгосрочную монетизацию.

Retention-динамика демонстрирует двухфазный lifecycle. Первая фаза — агрессивный ранний churn, где система теряет значительную часть недавно привлеченных клиентов. Вторая — фаза стабилизации, в которой оставшиеся пользователи начинают демонстрировать repeat purchasing behavior и более высокую интенсивность трат.

Ключевой вывод состоит в том, что качество retained аудитории со временем растет быстрее, чем сокращается ее объем. Это видно через отклонение между user retention и revenue retention, а также через устойчивый рост revenue per retained user. Фактически lifecycle-система постепенно самостоятельно отбирает более ценных и вовлеченных покупателей.

Одновременно изменчивость когорт указывает на наличие структурной неоднородности клиентской базы. Поздние lifecycle-стадии все чаще состоят из разных поведенческих архетипов: часть пользователей остается эпизодическими покупателями, тогда как другая формирует долгосрочные модели привычного потребления. Это особенно характерно для аптечного сегмента, где присутствуют как разовые покупки, так и хроническое регулярное потребление.

Дополнительно cumulative LTV curves показывают, что даже относительно небольшое улучшение retention на ранних этапах может иметь непропорциональное долгосрочное влияние на revenue за счет накопительного эффекта монетизации.

### Влияние на бизнес

Текущая retention-модель создает высокую зависимость роста от эффективности управления на ранних этапах жизненного цикла. Если первоначальный churn сохранится на текущем уровне, бизнесу потребуется постоянно компенсировать потери за счет трат на привлечение (acquisition), что потенциально ухудшает unit экономики и повышает CAC чувствительность.

Одновременно анализ показывает сильный потенциал retention-led монетизации. Удержанные пользователи демонстрируют рост customer value во времени, что делает долгосрочный retention значительно более экономически эффективным источником роста revenue, чем постоянное расширение воронки привлечения (acquisition funnel).

Рост RPRU среди retained клиентов также снижает риск обвала revenue при умеренном снижении общей клиентской базы. Монетизация все чаще концентрируется внутри loyal/high-frequency клиентов, что повышает устойчивость потока revenue.

Однако высокая вариативность когорт на поздних стадиях создает риск концентрации revenue внутри ограниченного числа активных покупателей. В случае ухудшения retention среди этих сегментов бизнес может столкнуться ускоренным ухудшением LTV.

### Ограничения анализа

- Анализ основан на cohort-based observational data и не позволяет однозначно интерпретировать причинность изменений в поведении.  
- Часть поздних lifecycle периодов содержит ограниченное количество зрелых когорт, что повышает изменчивость и снижает статистическую устойчивость long-tail наблюдений.  
- Retention curves не учитывают внешние факторы: сезонность спроса, изменения ассортимента, промо-активности и макроэкономические эффекты.  
- Revenue retention интерпретируется на агрегированном cohort-уровне и не отражает индивидуальную неоднородность клиентов внутри когорт.  
- Дополнительно cohort framework фиксирует acquisition month как старт lifecycle, однако не учитывает возможную динамику реактивации после длительных периодов неактивности.  

### Приоритетные направления работы  

1. Сфокусировать retention-стратегию на первых 30–60 днях lifecycle, где концентрируется основной churn-риск и формируется долгосрочная клиентская траектория.
2. Построить early-repeat-purchase CRM flows: персонализированные напоминания, механики восполнения, триггеры лояльности и кампании по вовлечению клиентов после покупки.  
3. Выделить и отдельно анализировать high-LTV retained cohorts для понимания моделей поведения, формирующих долгосрочный рост монетизации.  
4. Запустить сегментацию между эпизодическими и постоянными покупателями для более точного управления retention инвестиций и CRM приоритезаций.  
5. Дополнительно исследовать драйвера роста RPRU: увеличение частоты покупок, размера покупательской корзины, расширение категории или эффектов изменения цен и ассортимента.  
6. Построить predictive retention/LTV model для раннего определения пользователей с высокой вероятностью перехода в loyal сегмент повторных покупателей.  
7. Провести контрольные эксперименты на ранних этапах жизненного цикла и оценить влияние на Month-1 retention, коэффициент повторных покупок и долгосрочный cohort LTV.
