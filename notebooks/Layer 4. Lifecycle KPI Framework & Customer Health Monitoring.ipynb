# Layer 4. Lifecycle KPI Framework & Customer Health Monitoring

**Бизнес-контекст**

Для эффективного управления клиентской базой недостаточно только прогнозировать вероятность оттока отдельных пользователей или анализировать текущее распределение клиентов по жизненному циклу.

Компании необходима система непрерывного мониторинга customer health, позволяющая одновременно оценивать текущее состояние клиентского портфеля, выявлять ранние признаки ухудшения retention dynamics и отслеживать изменения структуры выручки во времени.

В условиях высокой конкуренции и снижения потребительской активности аптечной сети важно понимать не только где концентрируется текущая клиентская ценность, но и как меняется устойчивость клиентской базы, скорость восстановления клиентов и качество удержания на различных этапах жизненного цикла.

Для решения этой задачи был разработан Lifecycle KPI Framework — система мониторинга клиентского здоровья, объединяющая lifecycle segmentation, behavioral analytics, revenue monitoring и predictive churn scoring в единую модель оценки customer health.

**Основная задача раздела**  

Построить систему lifecycle-monitoring метрик, позволяющую:  
- отслеживать состояние клиентской базы по lifecycle stages;  
- анализировать retention, churn, reactivation и resurrection dynamics;  
- оценивать устойчивость клиентской базы и стабильность lifecycle structure;  
- выявлять изменения customer health во времени;  
- анализировать распределение выручки между сегментами;   
- выявлять revenue concentration и revenue-at-risk;  
- определять наиболее приоритетные сегменты для retention-инвестиций;  
- интегрировать predictive churn scores в систему бизнес-мониторинга.

**Гипотезы**

1. Loyal-клиенты формируют непропорционально большую долю общей выручки компании.  
2. At Risk-сегмент концентрирует значительную часть потенциальных потерь revenue-at-risk.  
3. Средний churn-risk систематически возрастает по мере деградации lifecycle stage.  
4. Даже при относительно небольшом размере сегмента отдельные группы клиентов могут обладать критически высокой экономической значимостью.  
5. Интеграция predictive churn scoring в lifecycle monitoring позволяет более точно определять приоритеты retention-стратегии.  
6. По мере ухудшения структуры жизненного цикла показатель Lifecycle Stability Index будет снижаться.  
7. Стабильность Healthy-сегментов (Active + Loyal) является более чувствительным индикатором ухудшения customer health, чем традиционные retention-метрики.  

**Методология**

Анализ реализован на основе snapshot-based lifecycle monitoring framework.  
Для каждого временного snapshot рассчитываются behavioral, revenue и predictive metrics по lifecycle stages.  
В качестве основы используются результаты predictive churn-модели, включая:  
- churn probability;
- behavioral features;
- customer value indicators.

Клиенты классифицируются по стадиям жизненного цикла:
- New  
- Active  
- Loyal  
- At Risk  
- Churn  

Для каждого lifecycle stage рассчитываются:

- **Customer Metrics**
    - количество клиентов;
    - доля клиентской базы;
    - распределение пользователей между стадиями.
- **Revenue Metrics**
    - total revenue;
    - average revenue per customer;
    - revenue share;
    - revenue concentration.
- **Behavioral Metrics**
    - average purchase frequency;
    - average recency;
    - customer activity dynamics.
- **Predictive Metrics**
    - average churn probability;
    - revenue-at-risk;
    - priority score.

**Дополнительно реализованы временные индикаторы устойчивости клиентской базы**  
Для каждого периода рассчитываются:  

- Lifecycle Stability Index  
Доля клиентов, сохранивших свою lifecycle stage между соседними snapshot-периодами.  
Метрика отражает общую стабильность структуры клиентской базы и позволяет оценивать интенсивность lifecycle-миграций.  

- Healthy Stability Index  
Доля клиентов из здоровых сегментов (Active и Loyal), сохранивших свое состояние между периодами.  
Метрика используется как индикатор устойчивости наиболее ценной части клиентской базы и позволяет выявлять ранние признаки ухудшения customer health.

## Lifecycle KPI Aggregation Framework

После построения predictive churn-модели и определения приоритетных клиентов для retention-интервенций следующим шагом является переход от управления отдельными пользователями к мониторингу состояния клиентского портфеля в целом. Для принятия регулярных управленческих решений бизнесу необходима система агрегированных показателей, позволяющая отслеживать динамику удержания, восстановления клиентов, стабильность выручки, устойчивость lifecycle-сегментов и общее состояние клиентской базы во времени.

Цель анализа — сформировать единый Lifecycle KPI Framework и оценить, как изменяются ключевые показатели customer health между последовательными lifecycle-срезами. Дополнительно проверяется гипотеза о том, что изменения структуры клиентской базы отражаются не только в churn и retention-метриках, но также в уровне устойчивости жизненного цикла клиентов, способности системы удерживать здоровые сегменты и динамике удерживаемой выручки.

Для анализа используется snapshot-based lifecycle monitoring approach. Клиенты классифицируются по стадиям жизненного цикла на последовательных временных срезах, после чего рассчитываются ключевые показатели портфельного уровня: Retention Rate, Churn Rate, Reactivation Rate, Resurrection Rate, Revenue Retention, Expansion Revenue и Contraction Revenue.

Дополнительно рассчитываются два интегральных индикатора устойчивости клиентской базы:  
• Lifecycle Stability Index — доля клиентов, сохранивших свое lifecycle-состояние между соседними snapshot-периодами;  
• Healthy Lifecycle Stability Index — доля клиентов из здоровых сегментов (Active и Loyal), которые остаются в этих состояниях без деградации жизненного цикла.  

В отличие от традиционных retention-показателей данные метрики позволяют оценивать не только факт удержания клиентов, но и структурную устойчивость клиентского портфеля, а также способность бизнеса сохранять наиболее ценные сегменты пользовательской базы.

Ожидается, что по мере развития жизненного цикла будут проявляться различия между динамикой клиентского удержания, устойчивостью healthy-сегментов и динамикой удержания выручки. Если Revenue Retention и Healthy Lifecycle Stability остаются высокими при умеренном уровне клиентского churn, это может свидетельствовать о сохранении стабильного ядра клиентов, формирующего основную часть доходов компании. Напротив, одновременное ухудшение retention, stability и рост Contraction Revenue будут выступать ранними индикаторами ухудшения customer health и накопления revenue-at-risk внутри клиентского портфеля.

Для формирования системы непрерывного мониторинга клиентского здоровья рассчитаем ключевые lifecycle, stability и revenue KPI между последовательными snapshot-периодами и оценим динамику удержания клиентов, устойчивости клиентской базы и изменения клиентской ценности во времени.


```python
# =====================================================
# IMPORTS
# =====================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns



# =====================================================
# LOAD DATA
# =====================================================

df = pd.read_csv('no_outliers_freq.csv')

df['datetime'] = pd.to_datetime(df['datetime'])


# =====================================================
# SNAPSHOT DATES
# =====================================================

snapshot_dates = pd.to_datetime([
    '2021-12-31',
    '2022-01-31',
    '2022-02-28',
    '2022-03-31',
    '2022-04-30',
    '2022-05-31'
])

# =====================================================
# ALL CUSTOMERS
# =====================================================

all_customers = pd.DataFrame({
    'card': df['card'].unique()
})

# =====================================================
# FIRST PURCHASE
# =====================================================

first_purchase = (

    df.groupby('card')['datetime']
    .min()
    .reset_index()

)

first_purchase.columns = [
    'card',
    'first_purchase_date'
]

# =====================================================
# SNAPSHOT FUNCTION
# =====================================================

def create_snapshot(data, snapshot_date):

    # -------------------------------------------------
    # HISTORY BEFORE SNAPSHOT
    # -------------------------------------------------

    history = data[
        data['datetime'] <= snapshot_date
    ].copy()

    # -------------------------------------------------
    # 6-MONTH WINDOW
    # -------------------------------------------------

    start_date = (
        snapshot_date
        - pd.DateOffset(months=6)
    )

    window = history[
        history['datetime'] >= start_date
    ].copy()

    # -------------------------------------------------
    # RECENCY
    # -------------------------------------------------

    recency = (

        history.groupby('card')['datetime']
        .max()
        .reset_index()

    )

    recency['Recency'] = (
        snapshot_date
        - recency['datetime']
    ).dt.days

    recency = recency[
        ['card', 'Recency']
    ]

    # -------------------------------------------------
    # FREQUENCY + MONETARY
    # -------------------------------------------------

    fm = (

        window.groupby('card')
        .agg({
            'doc_id': 'nunique',
            'summ_with_disc': 'sum'
        })
        .reset_index()

    )

    fm.columns = [
        'card',
        'Frequency',
        'Monetary'
    ]

    # -------------------------------------------------
    # MERGE
    # -------------------------------------------------

    rfm = all_customers.merge(
        recency,
        on='card',
        how='left'
    )

    rfm = rfm.merge(
        fm,
        on='card',
        how='left'
    )

    rfm = rfm.merge(
        first_purchase,
        on='card',
        how='left'
    )

    # -------------------------------------------------
    # KEEP VALID CUSTOMERS
    # -------------------------------------------------

    rfm = rfm[
        rfm['first_purchase_date']
        <= snapshot_date
    ]

    rfm = rfm[
        rfm['Recency'].notna()
    ]

    # -------------------------------------------------
    # FILL MISSING
    # -------------------------------------------------

    rfm['Frequency'] = (
        rfm['Frequency']
        .fillna(0)
    )

    rfm['Monetary'] = (
        rfm['Monetary']
        .fillna(0)
    )

    # -------------------------------------------------
    # CUSTOMER AGE
    # -------------------------------------------------

    rfm['Customer_Age'] = (

        snapshot_date
        - rfm['first_purchase_date']

    ).dt.days

    # =================================================
    # LIFECYCLE STAGES
    # =================================================

    conditions = [

        # NEW
        (
            rfm['Customer_Age'] <= 31
        ),

        # LOYAL
        (
            (rfm['Frequency'] >= 4)
            &
            (rfm['Recency'] <= 88)
        ),

        # ACTIVE
        (
            rfm['Recency'] <= 88
        ),

        # AT RISK
        (
            rfm['Recency']
            .between(89, 184)
        ),

        # CHURN
        (
            rfm['Recency'] > 184
        )
    ]

    stages = [
        'New',
        'Loyal',
        'Active',
        'At Risk',
        'Churn'
    ]

    rfm['Lifecycle_Stage'] = np.select(
        conditions,
        stages,
        default='Unknown'
    )

    rfm['Snapshot_Date'] = snapshot_date

    return rfm

# =====================================================
# BUILD SNAPSHOTS
# =====================================================

snapshots = []

for snapshot_date in snapshot_dates:

    snap = create_snapshot(
        df,
        snapshot_date
    )

    snapshots.append(snap)

all_snapshots = pd.concat(
    snapshots,
    ignore_index=True
)

# =====================================================
# BUILD TRANSITIONS
# =====================================================

transition_dfs = []

for i in range(len(snapshot_dates)-1):

    current_date = snapshot_dates[i]
    next_date = snapshot_dates[i+1]

    # -------------------------------------------------
    # CURRENT SNAPSHOT
    # -------------------------------------------------

    current_snapshot = all_snapshots[

        all_snapshots['Snapshot_Date']
        == current_date

    ][[
        'card',
        'Lifecycle_Stage',
        'Monetary'
    ]]

    # -------------------------------------------------
    # NEXT SNAPSHOT
    # -------------------------------------------------

    next_snapshot = all_snapshots[

        all_snapshots['Snapshot_Date']
        == next_date

    ][[
        'card',
        'Lifecycle_Stage',
        'Monetary'
    ]]

    # -------------------------------------------------
    # MERGE
    # -------------------------------------------------

    merged = current_snapshot.merge(

        next_snapshot,

        on='card',

        how='inner',

        suffixes=('_from', '_to')

    )

    # -------------------------------------------------
    # REVENUE CHANGE
    # -------------------------------------------------

    merged['Revenue_Change'] = (

        merged['Monetary_to']
        -
        merged['Monetary_from']

    )

    # -------------------------------------------------
    # PERIOD
    # -------------------------------------------------

    merged['Period'] = (

        f"{current_date.date()}"
        +
        " -> "
        +
        f"{next_date.date()}"

    )

    transition_dfs.append(merged)

# =====================================================
# FINAL TRANSITIONS TABLE
# =====================================================

all_transitions = pd.concat(
    transition_dfs,
    ignore_index=True
)

# =====================================================
# KPI FRAMEWORK
# =====================================================

kpi_rows = []

for period in all_transitions['Period'].unique():

    df_period = all_transitions[
        all_transitions['Period'] == period
    ].copy()

    # =================================================
    # CUSTOMER COUNTS
    # =================================================

    total_customers = len(df_period)

    retained_customers = len(

        df_period[

            df_period['Lifecycle_Stage_to']
            .isin(['Active', 'Loyal'])

        ]

    )

    churned_customers = len(

        df_period[
            df_period['Lifecycle_Stage_to']
            == 'Churn'
        ]

    )

    reactivated_customers = len(

        df_period[

            (
                df_period['Lifecycle_Stage_from']
                .isin(['At Risk', 'Churn'])
            )

            &

            (
                df_period['Lifecycle_Stage_to']
                .isin(['Active', 'Loyal'])
            )

        ]

    )

    resurrected_customers = len(

        df_period[

            (
                df_period['Lifecycle_Stage_from']
                == 'Churn'
            )

            &

            (
                df_period['Lifecycle_Stage_to']
                .isin(['Active', 'Loyal'])
            )

        ]

    )

    # =================================================
    # REVENUE METRICS
    # =================================================

    start_revenue = (
        df_period['Monetary_from']
        .sum()
    )

    retained_revenue = (

        df_period[

            df_period['Lifecycle_Stage_to']
            .isin(['Active', 'Loyal'])

        ]['Monetary_to']
        .sum()

    )

    expansion_revenue = (

        df_period[
            df_period['Revenue_Change'] > 0
        ]['Revenue_Change']
        .sum()

    )

    contraction_revenue = abs(

        df_period[
            df_period['Revenue_Change'] < 0
        ]['Revenue_Change']
        .sum()

    )

    # =================================================
    # KPI CALCULATIONS
    # =================================================

    retention_rate = (
        retained_customers
        / total_customers
    )

    churn_rate = (
        churned_customers
        / total_customers
    )

    at_risk_base = len(

        df_period[

            df_period['Lifecycle_Stage_from']
            == 'At Risk'

        ]

    )

    reactivation_rate = (

        reactivated_customers
        /
        max(at_risk_base, 1)

    )

    churn_base = len(

        df_period[

            df_period['Lifecycle_Stage_from']
            == 'Churn'

        ]

    )

    resurrection_rate = (

        resurrected_customers
        /
        max(churn_base, 1)

    )

    revenue_retention = (

        retained_revenue
        /
        start_revenue

    )

    # =================================================
    # SAVE
    # =================================================

    kpi_rows.append({

        'Period': period,

        'Retention_Rate': retention_rate,

        'Churn_Rate': churn_rate,

        'Reactivation_Rate': reactivation_rate,

        'Resurrection_Rate': resurrection_rate,

        'Revenue_Retention': revenue_retention,

        'Expansion_Revenue': expansion_revenue,

        'Contraction_Revenue': contraction_revenue

    })

# =====================================================
# FINAL KPI TABLE
# =====================================================

kpi_df = pd.DataFrame(kpi_rows)

kpi_df = kpi_df.round(3)

# =====================================================
# LIFECYCLE STABILITY INDICES
# =====================================================

healthy_stages = [
    'Active',
    'Loyal'
]

stability_rows = []

for period in all_transitions['Period'].unique():

    temp = all_transitions[
        all_transitions['Period'] == period
    ].copy()

    # =================================================
    # OVERALL STABILITY
    # =================================================

    total_customers = len(temp)

    stable_customers = len(

        temp[

            temp['Lifecycle_Stage_from']
            ==
            temp['Lifecycle_Stage_to']

        ]

    )

    overall_stability = (

        stable_customers
        /
        max(total_customers, 1)

    )

    # =================================================
    # HEALTHY STABILITY
    # =================================================

    healthy_base = temp[

        temp['Lifecycle_Stage_from']
        .isin(healthy_stages)

    ]

    stable_healthy = healthy_base[

        healthy_base['Lifecycle_Stage_from']
        ==
        healthy_base['Lifecycle_Stage_to']

    ]

    healthy_stability = (

        len(stable_healthy)

        /
        max(len(healthy_base), 1)

    )

    # =================================================
    # SAVE
    # =================================================

    stability_rows.append({

        'Period': period,

        'Lifecycle_Stability_Index':
        overall_stability,

        'Healthy_Stability_Index':
        healthy_stability

    })

# =====================================================
# FINAL TABLE
# =====================================================

stability_df = pd.DataFrame(
    stability_rows
)

stability_df = (
    stability_df
    .round(3)
)

# =====================================================
# MERGE KPI + STABILITY
# =====================================================

kpi_df = kpi_df.merge(

    stability_df,

    on='Period',

    how='left'

)

kpi_df = kpi_df.round(3)

print("="*60)
print("LIFECYCLE KPI & STABILITY FRAMEWORK")
print("="*60)

print(kpi_df)
```

## Lifecycle KPI Trends Visualization

На данном этапе проанализируем динамику ключевых lifecycle, stability и revenue-метрик между последовательными snapshot-периодами. Если предыдущий этап был направлен на построение системы мониторинга клиентского здоровья, то теперь важно оценить, как изменяется состояние клиентской базы во времени, насколько устойчивыми остаются основные lifecycle-сегменты и как эти изменения влияют на финансовые показатели бизнеса.

## Lifecycle Stage KPI Dashboard

На данном этапе исследуется текущее состояние клиентского портфеля на уровне lifecycle-сегментов. После анализа общей динамики retention и churn важно понять, какие группы клиентов формируют основную часть выручки компании, как различается их поведенческий профиль и где концентрируются ключевые бизнес-риски.

Цель анализа — определить экономическую и стратегическую значимость отдельных lifecycle stages, а также выявить сегменты, сочетающие высокий вклад в выручку с повышенной вероятностью потери клиентов. Особый интерес представляют группы, в которых одновременно наблюдаются высокая ценность для бизнеса и высокий churn-risk, поскольку именно такие сегменты формируют основную часть потенциальных потерь будущей выручки.

Для анализа используется актуальный snapshot клиентской базы, дополненный результатами модели прогнозирования оттока. Для каждого lifecycle stage рассчитываются показатели клиентского масштаба, выручки, покупательской активности и прогнозного риска. Помимо традиционных KPI дополнительно оцениваются Revenue Share, Revenue Index и Revenue-at-Risk, позволяющие связать поведенческие характеристики клиентов с их экономическим вкладом и вероятностью будущих потерь.

Ожидается, что наиболее зрелые и активные сегменты будут концентрировать значительную долю общей выручки, тогда как стадии деградации жизненного цикла продемонстрируют повышенные значения churn-risk и непропорционально высокий объем revenue-at-risk. Подобный анализ позволяет перейти от мониторинга клиентской базы в целом к приоритизации retention-инициатив на основе совокупности риска и бизнес-ценности сегмента.

Рассчитаем ключевые behavioral, revenue и predictive KPI для каждого lifecycle stage и сформируем сегментный dashboard клиентского здоровья.

```python
# =====================================================
# LOAD DATA
# =====================================================

# full snapshot dataset
#all_snapshots = pd.read_csv(
#    'all_snapshots.csv'
#)

# scored dataset from ML notebook
results = pd.read_csv(
    'customer_churn_scores.csv'
)

# =====================================================
# DATETIME
# =====================================================

all_snapshots['Snapshot_Date'] = pd.to_datetime(
    all_snapshots['Snapshot_Date']
)

results['Snapshot_Date'] = pd.to_datetime(
    results['Snapshot_Date']
)

# =====================================================
# LATEST SNAPSHOT DATE
# =====================================================

# IMPORTANT:
# latest date should come from FULL snapshot table,
# because it contains the complete lifecycle base

latest_snapshot_date = (
    all_snapshots['Snapshot_Date']
    .max()
)

print("="*60)
print("LATEST SNAPSHOT DATE")
print("="*60)

print(latest_snapshot_date)

# =====================================================
# LATEST SNAPSHOT (FULL BASE)
# =====================================================

latest_snapshot = all_snapshots[

    all_snapshots['Snapshot_Date']
    ==
    latest_snapshot_date

].copy()

# =====================================================
# LATEST CHURN SCORES
# =====================================================

latest_scores = results[

    results['Snapshot_Date']
    ==
    latest_snapshot_date

][[
    'card',
    'Churn_Probability'
]].copy()

# =====================================================
# MERGE CHURN SCORES
# =====================================================

kpi_df = latest_snapshot.merge(

    latest_scores,

    on='card',

    how='left'

)

# =====================================================
# FILL MISSING SCORES
# =====================================================

# if some customers do not have prediction
# (rare edge case)

kpi_df['Churn_Probability'] = (
    kpi_df['Churn_Probability']
    .fillna(0)
)

# =====================================================
# TOTALS
# =====================================================

total_customers = len(kpi_df)

total_revenue = (
    kpi_df['Monetary']
    .sum()
)

overall_avg_revenue = (
    kpi_df['Monetary']
    .mean()
)

# =====================================================
# STAGE KPI DASHBOARD
# =====================================================

stage_kpi = (

    kpi_df

    .groupby(
        'Lifecycle_Stage',
        observed=True
    )

    .agg({

        'card': 'count',

        'Monetary': [
            'sum',
            'mean'
        ],

        'Frequency': 'mean',

        'Recency': 'mean',

        'Churn_Probability': 'mean'

    })

)

# =====================================================
# FLATTEN COLUMNS
# =====================================================

stage_kpi.columns = [

    'Customers',

    'Total_Revenue',

    'Avg_Revenue',

    'Avg_Frequency',

    'Avg_Recency',

    'Avg_Churn_Risk'

]

stage_kpi = stage_kpi.reset_index()

# =====================================================
# CUSTOMER SHARE
# =====================================================

#stage_kpi['Customer_Share_%'] = (

#    stage_kpi['Customers']
#    /
#    total_customers
#    * 100

#)

# =====================================================
# REVENUE SHARE
# =====================================================

stage_kpi['Revenue_Share_%'] = (

    stage_kpi['Total_Revenue']
    /
    total_revenue
    * 100

)

# =====================================================
# REVENUE INDEX
# =====================================================

stage_kpi['Revenue_Index'] = (

    stage_kpi['Avg_Revenue']
    /
    overall_avg_revenue

)

# =====================================================
# REVENUE AT RISK
# =====================================================

stage_kpi['Revenue_At_Risk'] = (

    stage_kpi['Total_Revenue']
    *
    stage_kpi['Avg_Churn_Risk']

)

# =====================================================
# PRIORITY SCORE
# =====================================================

stage_kpi['Priority_Score'] = (

    stage_kpi['Revenue_Share_%']
    *
    stage_kpi['Avg_Churn_Risk']

)

# =====================================================
# ROUND
# =====================================================

stage_kpi = stage_kpi.round({

    'Total_Revenue': 0,
    'Avg_Revenue': 0,

    # behavioral metrics
    'Avg_Frequency': 1,
    'Avg_Recency': 0,

    # probabilities
    'Avg_Churn_Risk': 2,

    # shares
#    'Customer_Share_%': 1,
    'Revenue_Share_%': 1,

    # business indexes
    'Revenue_Index': 2,
    'Revenue_At_Risk': 0,
    'Priority_Score': 1

})

# =====================================================
# SORT
# =====================================================

stage_kpi = stage_kpi.sort_values(

    'Total_Revenue',
    ascending=False

)

# =====================================================
# OUTPUT
# =====================================================

print("\n")
print("="*70)
print("LIFECYCLE STAGE KPI DASHBOARD")
print("="*70)

print(stage_kpi)
```

**Key Findings**

Lifecycle-анализ показал высокую концентрацию выручки в сегменте Loyal: 904 клиента (≈16% базы) формируют 55.4% всей выручки и имеют средний чек в 3.5 раза выше среднего по базе. Одновременно наибольший объем revenue-at-risk сосредоточен не в Loyal, а в сегментах Active и At Risk — 1.39 млн и 1.35 млн соответственно. Средний churn-risk последовательно растет по мере деградации жизненного цикла: 0.16 у Loyal, 0.52 у Active, 0.68 у At Risk и 0.95 у Churn.

Клиентская экономика сети зависит от относительно небольшого ядра лояльных покупателей, однако основной риск потери будущей выручки формируется на более ранних стадиях жизненного цикла. Active и At Risk уже генерируют существенную долю оборота, но демонстрируют значительно более высокий риск ухода, что делает их ключевой зоной для retention-инвестиций. Priority Score подтверждает, что максимальный бизнес-эффект следует ожидать не от работы с уже ушедшими клиентами, а от предотвращения дальнейшей деградации активной клиентской базы.

Возможной причиной может быть недостаточная интенсивность повторного вовлечения после первых покупок и отсутствие механизмов перевода клиентов из Active в Loyal. Предположительно часть клиентов сохраняет эпизодическое потребление и постепенно переходит в стадию риска до формирования устойчивой покупательской привычки.

Основная задача retention-стратегии — не реактивация Churn-клиентов, а ускорение продвижения клиентов по жизненному циклу и снижение потерь на переходах Active → At Risk и At Risk → Churn.

**Prioritized Actions**  

Приоритет №1 — запуск ранних retention-кампаний для сегмента Active с высоким churn-score.  
Приоритет №2 — персонализированные win-back сценарии для At Risk клиентов до наступления фактического оттока.  
Приоритет №3 — программы лояльности и cross-sell механики для ускорения перехода Active → Loyal.  

Снижение churn среди Active и At Risk сегментов способно защитить крупнейший объем revenue-at-risk, повысить Revenue Retention и увеличить долю клиентов, достигающих наиболее прибыльной стадии Loyal.

Ключевым фактором роста становится не столько масштабирование клиентской базы, сколько сохранение и развитие наиболее ценных lifecycle-сегментов. Для бизнеса основной риск связан с потерей части Loyal-клиентов и дальнейшим переходом At Risk-аудитории в полноценный churn, поскольку именно эти группы концентрируют большую часть денежных потоков компании.

**Prioritized Actions**

1. Приоритизировать удержание Loyal-клиентов как основного источника выручки.
2. Развернуть проактивные retention-механики для сегмента At Risk до фактического оттока.
3. Отслеживать динамику revenue concentration во времени как индикатор устойчивости клиентского портфеля.
4. Оценить долю Loyal-клиентов, переходящих в At Risk, как ключевой leading indicator будущего снижения выручки.

Снижение скорости деградации Loyal-клиентов и повышение эффективности реактивации At Risk-аудитории позволит сохранить наиболее доходную часть клиентской базы, сократить revenue leakage и повысить долгосрочную устойчивость денежных потоков компании.

## Churn Risk Distribution Across Lifecycle Stages

На данном этапе анализа переходим от оценки структуры выручки к исследованию распределения риска оттока внутри клиентского портфеля. Если revenue concentration показывает, где сосредоточена текущая экономическая ценность бизнеса, то анализ churn risk позволяет понять, насколько устойчивы эти источники выручки и какие lifecycle-сегменты находятся под наибольшей угрозой потери.

Цель анализа — оценить, как вероятность оттока распределяется между стадиями жизненного цикла клиентов, и проверить, соответствует ли фактическое поведение пользователей логике lifecycle-модели. Дополнительно важно определить, какие сегменты требуют приоритетного внимания со стороны retention-стратегий и насколько рано компания способна выявлять признаки деградации клиентской активности.

Для анализа используются средние значения churn probability, рассчитанные predictive-моделью для каждого lifecycle stage. Такой подход позволяет объединить историческое поведение клиентов и прогнозный риск в единую систему customer health monitoring, а также сопоставить традиционную сегментацию с вероятностью будущего оттока.

Ожидается, что по мере продвижения клиентов от Active и Loyal к At Risk и Churn будет наблюдаться последовательный рост вероятности оттока. Если такая зависимость подтверждается, это будет свидетельствовать о корректной работе как lifecycle-сегментации, так и predictive churn framework. В противном случае это может указывать на наличие скрытых поведенческих паттернов, не отражаемых текущими правилами сегментации.

Рассмотрим распределение среднего churn risk между lifecycle stages и оценим, какие сегменты формируют основные зоны риска для будущего удержания клиентской базы.

```python
# =====================================================
# CHURN RISK VISUALIZATION
# =====================================================

plt.figure(figsize=(10,6))

plt.bar(

    stage_kpi['Lifecycle_Stage'],

    stage_kpi['Avg_Churn_Risk']

)

plt.title(
    'Average Churn Risk by Lifecycle Stage'
)

plt.ylabel('Churn Probability')

plt.show()
```

**Key Findings**

Средняя вероятность оттока демонстрирует последовательный рост по мере деградации жизненного цикла клиента. Минимальный риск наблюдается в сегменте Loyal (16%), тогда как для Active клиентов он составляет 52%, для At Risk — 68%, а для Churn достигает 95%. Сегмент New демонстрирует промежуточный уровень риска (45%), превышающий показатель наиболее лояльной аудитории почти в три раза.

Результаты подтверждают, что lifecycle-сегментация корректно отражает фактическое состояние клиентской базы и согласуется с прогнозами churn-модели. По мере снижения вовлеченности клиентов вероятность потери систематически возрастает, что позволяет рассматривать lifecycle stage как операционный индикатор customer health. Особенно важно, что высокий риск наблюдается уже на стадии Active, где клиенты продолжают совершать покупки и генерировать существенную часть выручки. Это означает, что значительная доля будущего оттока формируется задолго до перехода пользователей в сегмент At Risk.

Предположительно высокая вероятность оттока среди Active-клиентов связана с неоднородностью данного сегмента. Часть пользователей формально остается активной, однако уже демонстрирует ранние признаки снижения частоты покупок и вовлеченности, которые фиксируются predictive-моделью раньше, чем lifecycle-правилами сегментации.

Текущая система мониторинга показывает, что управление удержанием не должно ограничиваться клиентами, уже попавшими в At Risk. Значительная часть потенциальных потерь формируется на более ранних стадиях жизненного цикла, где стоимость удержания обычно ниже, а вероятность успешной реактивации выше.

**Prioritized Actions**

1. Использовать churn score как ранний сигнал риска внутри сегмента Active.  
2. Выделить отдельные retention-сценарии для Active-клиентов с повышенной вероятностью оттока.  
3. Усилить мониторинг переходов Active → At Risk как ключевого этапа формирования будущего churn.  
4. Интегрировать churn probability в CRM-приоритизацию вместо использования только lifecycle-статуса.  

Более раннее выявление клиентов с ухудшающимся поведением позволит снизить скорость накопления At Risk-аудитории, повысить эффективность retention-коммуникаций и сократить будущие потери выручки за счет превентивного вмешательства до фактической деградации клиентской активности.  

## Revenue-at-Risk Prioritization Matrix

На данном этапе анализа переходим от изолированной оценки выручки и риска к комплексной оценке экономической значимости lifecycle-сегментов. Ранее были отдельно рассмотрены вклад сегментов в выручку и уровень churn-risk, однако для принятия retention-решений важно анализировать эти факторы совместно.

Цель анализа — определить, какие сегменты одновременно обладают высокой коммерческой ценностью и повышенной вероятностью оттока. Фактически мы проверяем, где сосредоточен наибольший объем потенциально уязвимой выручки и какие группы клиентов должны получать приоритетное внимание со стороны CRM и retention-команд.

Для этого используем матрицу Revenue vs Risk, в которой каждый lifecycle-сегмент оценивается по двум ключевым измерениям: экономической значимости (объем генерируемой выручки) и среднему прогнозируемому риску оттока. Дополнительно учитывается размер сегмента, что позволяет анализировать не только качество клиентской базы, но и масштаб потенциального бизнес-эффекта.

Ожидается, что сегменты будут распределены по различным зонам риска и ценности. Особый интерес представляют группы, сочетающие высокий вклад в выручку с повышенным churn-risk, поскольку именно они формируют основную часть revenue-at-risk и являются наиболее перспективными объектами для retention-инвестиций.

Построим матрицу Revenue vs Risk и определим приоритетные lifecycle-сегменты с точки зрения управления клиентской базой и защиты будущей выручки.

```python
# =====================================================
# REVENUE VS RISK MATRIX
# =====================================================

plt.figure(figsize=(10,6))

plt.scatter(

    stage_kpi['Avg_Churn_Risk'],

    stage_kpi['Total_Revenue'],

    s=stage_kpi['Customers'] * 2,

    alpha=0.7

)

for _, row in stage_kpi.iterrows():

    plt.text(

        row['Avg_Churn_Risk'],

        row['Total_Revenue'],

        row['Lifecycle_Stage']

    )

plt.xlabel('Average Churn Risk')

plt.ylabel('Total Revenue')

plt.title(
    'Lifecycle Revenue vs Risk Matrix'
)

plt.show()
```

**Key Findings**

Матрица Revenue vs Risk показала существенную неоднородность lifecycle-сегментов по сочетанию экономической ценности и риска оттока. Сегмент Loyal формирует основную часть выручки клиентской базы (6.46 млн руб.) при минимальном среднем churn-risk (16%). Сегменты Active и At Risk генерируют сопоставимые объемы выручки (2.66 млн и 1.98 млн руб. соответственно), однако характеризуются значительно более высоким прогнозируемым риском оттока — 52% и 68%. При этом именно Active является крупнейшим сегментом по численности клиентов (1 612 пользователей).

Полученные результаты показывают, что основной риск для бизнеса сосредоточен не в уже ушедших клиентах и не в наиболее лояльной аудитории, а в промежуточных lifecycle-стадиях. Несмотря на то что Loyal-сегмент обеспечивает большую часть текущей выручки, его поведенческий профиль остается относительно стабильным. Напротив, Active и At Risk одновременно обладают существенной экономической значимостью и повышенной вероятностью оттока, что делает их ключевыми источниками будущих потерь выручки.

С точки зрения customer lifecycle economics именно эти сегменты определяют качество retention-системы. Потеря даже небольшой доли клиентов из данных групп может оказать более заметное влияние на будущие денежные потоки, чем дальнейшая работа с уже ушедшими пользователями.

Возможной причиной такой структуры может быть отсутствие эффективного механизма перевода клиентов из стадии регулярной активности в устойчивую лояльность. Предположительно значительная часть пользователей совершает несколько покупок, но не формирует долгосрочную привычку взаимодействия с аптечной сетью. Для At Risk-сегмента дополнительным фактором может выступать снижение частоты покупок без своевременного CRM-вмешательства на ранних этапах деградации активности.

Основным объектом retention-стратегии должны стать не клиенты с максимальным churn-risk, а сегменты, находящиеся на пересечении высокой выручки и высокого риска. Матрица показывает, что оптимизация удержания Active и At Risk потенциально обеспечивает наибольшую отдачу на инвестиции в CRM и персонализированные коммуникации.

**Prioritized Actions**

1. внедрить ранние триггеры удержания для Active-клиентов до перехода в стадию At Risk.  
2. разработать отдельные win-back сценарии для At Risk-сегмента с персонализированными предложениями и стимулированием повторной покупки.  
3. масштабировать практики и поведенческие паттерны Loyal-клиентов на менее зрелые сегменты через программы лояльности и механики повышения частоты покупок.  
4. использовать churn-score как критерий приоритизации CRM-кампаний внутри Active и At Risk сегментов.  

Фокус на сегментах с максимальной концентрацией выручки и риска позволяет снизить объем потенциальных потерь revenue-at-risk, повысить retention наиболее ценной части клиентской базы и увеличить долю клиентов, переходящих из Active в Loyal. В долгосрочной перспективе это способствует росту customer lifetime value, стабилизации выручки и повышению эффективности retention-бюджета.

Ожидаемый эффект: рост LTV и снижение зависимости бизнеса от привлечения новых клиентов.  

**3. Мониторинг Revenue-at-Risk**  
Интегрировать Revenue-at-Risk, Revenue Retention и Lifecycle Migration в регулярную управленческую отчетность.  
Ожидаемый эффект: более раннее обнаружение ухудшения customer health и повышение качества принятия решений.  

**4. Segment-Level Retention эксперименты**  
Запустить серию retention-экспериментов для сегментов Active и At Risk с последующей оценкой влияния на churn, retention и revenue retention.  
Ожидаемый эффект: рост эффективности retention-бюджета и повышение ROI CRM-активностей.  

**5. Мониторинг Lifecycle Stability**  
Интегрировать Lifecycle Stability Index и Healthy Lifecycle Stability Index в регулярную систему customer health monitoring.  
Ожидаемый эффект:  
- раннее выявление изменений в структуре клиентской базы;  
- мониторинг устойчивости ключевых lifecycle-сегментов;   
- своевременное обнаружение ускорения негативных migration-паттернов.  

### Итоговый бизнес-вывод 

Анализ показывает, что главный резерв роста аптечной сети находится не в расширении клиентской базы, а в повышении эффективности управления жизненным циклом существующих клиентов. Ключевой задачей становится защита выручки на стадиях Active и At Risk, где сегодня концентрируется основная часть будущих потерь и одновременно находится наибольший потенциал для роста customer lifetime value.
