# Layer 1. Lifecycle Transition Analytics 

**Бизнес-контекст**  

Для аптечной сети критически важно понимать не только текущую структуру клиентской базы, но и динамику переходов пользователей между различными стадиями жизненного цикла. Проведенный ранее RFM-анализ позволил выделить ключевые lifecycle-сегменты клиентской базы и определить различия между пользователями по уровню активности, лояльности и покупательской ценности.

Однако статическая сегментация фиксирует состояние клиентов лишь в конкретный момент времени и не позволяет понять, каким образом пользователи перемещаются между lifecycle-состояниями, какие сегменты являются источниками будущего роста или потерь и как изменение клиентского поведения влияет на динамику выручки.

Для ответа на эти вопросы необходим переход от статического анализа сегментов к анализу динамики жизненного цикла клиентов.

В условиях высокой конкуренции и снижения покупательской активности компании необходимо перейти от разового анализа сегментов к динамическому lifecycle-подходу, который позволяет:
- отслеживать эволюцию клиентской базы во времени,  
- выявлять основные источники роста и потери revenue,  
- анализировать механики удержания и деградации пользователей,  
- оценивать вклад retention, reactivation и churn в изменение выручки,  
- определять наиболее критичные точки потери клиентов.  

Особую ценность представляет анализ customer transition dynamics, позволяющий оценивать не только вероятности переходов между lifecycle-состояниями, но и общее состояние (customer health) клиентской базы через динамику удержания, деградации и восстановления пользователей.

**Основная задача раздела**

Построить систему lifecycle transition analytics, позволяющую:  
- анализировать миграцию клиентов между lifecycle-сегментами,
- оценивать устойчивость клиентской базы,
- измерять retention и churn dynamics,
- декомпозировать изменение выручки по behavioral drivers,
- выявлять ключевые источники revenue growth и revenue loss,
- оценивать вклад acquisition, retention, expansion, contraction и churn в общую динамику бизнеса.

**Исследовательские вопросы**  

1. Насколько стабильны различные lifecycle-сегменты клиентской базы во времени?
2. Какие переходы между стадиями жизненного цикла происходят наиболее часто?
3. Какие lifecycle-состояния обладают наибольшим риском деградации и перехода в churn?
4. Какие пользовательские переходы формируют основной рост выручки?
5. Какие сегменты генерируют максимальные revenue losses?
6. Как распределяются денежные потоки между acquisition, retention, expansion, contraction и churn?
7. Насколько существенную роль играют:
    - реактивация клиентов,
    - удержание активной базы,
    - expansion за счет существующих клиентов
    в сравнении с привлечением новых пользователей?  
8. Улучшается или ухудшается состояние клиентской базы по мере развития жизненного цикла и как это отражается на структуре будущей выручки?  

**Гипотезы** 
1. По мере удаления клиента от стадии Loyal вероятность деградации жизненного цикла возрастает, а вероятность удержания и повторной активности снижается.  
2. Основные потери выручки формируются не только за счет полного churn, но и вследствие постепенного снижения покупательской активности активной базы.
3. Значительная часть роста revenue формируется за счет расширения трат (expansion) лояльных клиентов, а не исключительно за счет привлечения (acquisition) новых пользователей.
4. Реактивация клиентов из At Risk и Churn сегментов способна генерировать существенный вклад в восстановление выручки.
5. Lifecycle transitions обладают высокой предсказуемостью и могут использоваться как основа для построения retention-driven CRM-стратегии.


**Методология**

Анализ реализован в формате snapshot-based lifecycle analytics.  
Для каждого временного snapshot формируются пользовательские RFM-метрики:  
- Recency — давность последней покупки;
- Frequency — количество транзакций;
- Monetary — объем выручки;
- Customer Age — длительность отношений клиента с компанией.

Последовательность snapshot-периодов позволяет наблюдать эволюцию клиентской базы во времени и фиксировать изменения lifecycle-состояний для каждого клиента между соседними периодами наблюдения.

Далее пользователи классифицируются по lifecycle-стадиям:  
- New 
- Active
- Loyal
- At Risk
- Churn

Классификация строится на основе:  
- recency thresholds,
- transactional frequency,
- customer tenure.

После формирования snapshot-таблиц строится transition framework, позволяющий анализировать:  
- вероятности переходов между состояниями,
- стабильность retention,
- динамику churn,
- миграцию между lifecycle стадиями.  

Для оценки динамики выручки дополнительно рассчитываются:
- revenue transitions,
- revenue deltas,
- expansion/contraction effects,
- retained customer economics,
- revenue decomposition components.

В рамках анализа реализованы:
1. Transition matrices — вероятности переходов между lifecycle-состояниями;
2. Aggregated transition analysis — агрегированная оценка стабильности клиентской базы;
3. Sankey diagrams — визуализация customer flow и revenue flow между сегментами;
4. Revenue driver framework — классификация изменений выручки по драйверам:
    - Acquisition
    - Reactivation
    - Expansion
    - Contraction
    - Churn
6. Revenue decomposition model — декомпозиция изменения выручки на отдельные behavioral-компоненты;
7. Revenue transition heatmaps — анализ наиболее прибыльных и убыточных lifecycle transitions;
8. Retained customer economics — оценка expansion и contraction внутри удержанной клиентской базы.

**Логика исследования**

Анализ построен как последовательное исследование того, каким образом изменения клиентского поведения трансформируются в изменения выручки.
Общая логика исследования имеет следующий вид:
        
*Lifecycle Stage → Customer Transition → Retention / Churn Dynamics → Revenue Migration → Expansion / Contraction Effects → Revenue Growth & Revenue Loss*  

Таким образом, раздел позволяет перейти от описания структуры клиентской базы к пониманию механизмов формирования финансового результата компании.

**Бизнес-интерпретация**

Данный подход позволяет перейти от статического анализа клиентских сегментов к полноценному управлению customer lifecycle dynamics.

Компания получает возможность:

# Part 1. Customer Lifecycle Dynamics

##	Snapshots Generation

На данном этапе формируется аналитическая основа для исследования динамики клиентского жизненного цикла. Если традиционный RFM-анализ позволяет оценить состояние клиентской базы в конкретный момент времени, то для понимания механизмов удержания, деградации и восстановления клиентов необходимо перейти к анализу изменений между последовательными lifecycle-срезами.

Цель анализа — построить систему lifecycle snapshots и подготовить transition framework для оценки миграции клиентов между стадиями жизненного цикла. Проверяется гипотеза о том, что пользовательское поведение обладает устойчивыми закономерностями переходов между состояниями активности, а наблюдаемые изменения клиентской базы могут быть представлены как последовательность lifecycle-трансформаций, влияющих на будущий churn, retention и динамику выручки.

Для анализа используются последовательные временные snapshots клиентской базы. На каждом срезе рассчитываются ключевые поведенческие характеристики клиентов, включая давность последней покупки, частоту транзакций, накопленную выручку и длительность отношений с компанией. На основе этих показателей пользователи классифицируются по lifecycle stages (New, Active, Loyal, At Risk, Churn), после чего между соседними периодами формируются transition matrices, отражающие вероятности миграции клиентов между состояниями жизненного цикла.

Ожидается, что различные lifecycle-сегменты будут демонстрировать неодинаковую степень устойчивости. Если наиболее вовлеченные клиенты сохраняют высокий уровень удержания, а деградация концентрируется в отдельных стадиях жизненного цикла, это позволит локализовать ключевые точки потери клиентов и определить наиболее перспективные направления для retention-интервенций. Дополнительно сформированная transition framework станет основой для последующего анализа customer flow, revenue flow и декомпозиции изменений выручки по behavioral drivers.

Сформируем последовательные lifecycle snapshots клиентской базы и построим матрицы переходов между стадиями жизненного цикла для дальнейшего исследования customer transition dynamics.

```python
# =====================================================
# IMPORTS
# =====================================================

import pandas as pd
import numpy as np
import seaborn as sns
import plotly.graph_objects as go
import matplotlib.pyplot as plt

# =====================================================
# LOAD DATA
# =====================================================

df = pd.read_csv('no_outliers_freq.csv')

df['datetime'] = pd.to_datetime(df['datetime'])

# =====================================================
# SNAPSHOT DATES
# =====================================================

snapshot_dates = pd.to_datetime([
    '2021-12-31',
    '2022-01-31',
    '2022-02-28',
    '2022-03-31',
    '2022-04-30',
    '2022-05-31'
])

# =====================================================
# ALL CUSTOMERS
# =====================================================

all_customers = pd.DataFrame({
    'card': df['card'].unique()
})

# =====================================================
# FIRST PURCHASE TABLE
# =====================================================
first_purchase = df.groupby('card')['datetime'].min().reset_index()

first_purchase.columns = [
    'card',
    'first_purchase_date'
]

# =====================================================
# SNAPSHOT FUNCTION
# =====================================================

def create_snapshot(data, snapshot_date):

    # -----------------------------------------------
    # ALL HISTORY BEFORE SNAPSHOT
    # -----------------------------------------------

    history = data[
        data['datetime'] <= snapshot_date
    ].copy()

    # -----------------------------------------------
    # 6-MONTH WINDOW
    # -----------------------------------------------

    start_date = snapshot_date - pd.DateOffset(months=6)

    window = history[
        history['datetime'] >= start_date
    ].copy()

    # -----------------------------------------------
    # RECENCY (FULL HISTORY)
    # -----------------------------------------------

    recency = history.groupby('card').agg({
        'datetime': 'max'
    }).reset_index()

    recency['Recency'] = (
        snapshot_date - recency['datetime']
    ).dt.days

    recency = recency[['card', 'Recency']]

    # -----------------------------------------------
    # FREQUENCY + MONETARY (6-MONTH WINDOW)
    # -----------------------------------------------

    fm = window.groupby('card').agg({
        'doc_id': 'nunique',
        'summ_with_disc': 'sum'
    }).reset_index()

    fm.columns = [
        'card',
        'Frequency',
        'Monetary'
    ]

    # -----------------------------------------------
    # MERGE
    # -----------------------------------------------
    
    rfm = all_customers.merge(
        recency,
        on='card',
        how='left'
    )
    
    rfm = rfm.merge(
        fm,
        on='card',
        how='left'
    )
    
    rfm = rfm.merge(
        first_purchase,
        on='card',
        how='left'
    )


    rfm['Customer_Age'] = (
        snapshot_date - rfm['first_purchase_date']
    ).dt.days

    
    # -----------------------------------------------
    # KEEP ONLY ACQUIRED CUSTOMERS
    # -----------------------------------------------
    
    rfm = rfm[
        rfm['first_purchase_date'] <= snapshot_date
    ]
    
    # REMOVE INVALID USERS
    rfm = rfm[
        rfm['Recency'].notna()
    ]
    
    # -----------------------------------------------
    # FILL INACTIVE USERS
    # -----------------------------------------------
    
    rfm['Frequency'] = rfm['Frequency'].fillna(0)
    rfm['Monetary'] = rfm['Monetary'].fillna(0)
    
    # =================================================
    # LIFECYCLE STAGES
    # =================================================

    conditions = [
    
        # NEW
        (
            (rfm['Customer_Age'] <= 31)
        ),
    
        # LOYAL
        (
            (rfm['Frequency'] >= 4) &
            (rfm['Recency'] <= 88)
        ),
    
        # ACTIVE
        (
            (rfm['Recency'] <= 88)
        ),
    
        # AT RISK
        (
            (rfm['Recency'].between(89, 184))
        ),
    
        # CHURN
        (
            (rfm['Recency'] > 184)
        )
    ]
    stages = [
        'New',
        'Loyal',
        'Active',
        'At Risk',
        'Churn'
    ]

    rfm['Lifecycle_Stage'] = np.select(
        conditions,
        stages,
        default='Unknown'
    )
    
    # ==========================================
    # VALIDATION
    # ==========================================
    
    print("\n")
    print(f"SNAPSHOT: {snapshot_date}")
    
    print(
        rfm['Lifecycle_Stage']
        .value_counts(dropna=False)
    )
    
    print("\nUNKNOWN USERS:")
    
    print(
        rfm[
            rfm['Lifecycle_Stage'] == 'Unknown'
        ].head()
    )
    
    rfm['Snapshot_Date'] = snapshot_date
    
    return rfm

# =====================================================
# BUILD SNAPSHOTS
# =====================================================

snapshots = []

for snapshot_date in snapshot_dates:

    snap = create_snapshot(df, snapshot_date)

    snapshots.append(snap)

all_snapshots = pd.concat(snapshots)

# =====================================================
# TRANSITION MATRICES
# =====================================================

transition_matrices = []

for i in range(len(snapshot_dates)-1):

    current_date = snapshot_dates[i]
    next_date = snapshot_dates[i+1]

    current_snapshot = all_snapshots[
        all_snapshots['Snapshot_Date'] == current_date
    ][[
        'card',
        'Lifecycle_Stage',
        'Monetary'
    ]]
    
    next_snapshot = all_snapshots[
        all_snapshots['Snapshot_Date'] == next_date
    ][[
        'card',
        'Lifecycle_Stage',
        'Monetary'
    ]]

    merged = current_snapshot.merge(
        next_snapshot,
        on='card',
        how='inner',
        suffixes=('_from', '_to')
    )

    matrix = pd.crosstab(
        merged['Lifecycle_Stage_from'],
        merged['Lifecycle_Stage_to'],
        normalize='index'
    )

    transition_matrices.append({
        'period': f"{current_date.date()} -> {next_date.date()}",
        'matrix': matrix,
        'raw_data': merged
    })

```

*Этапы жизненного цикла (Lifecycle Stages) определены по принципу MECE (Mutually Exclusive, Collectively Exhaustive), обеспечивая взаимное исключение и полноту охвата клиентской базы. Каждый клиент может принадлежать только одной стадии на конкретную дату наблюдения. Дополнительно реализована процедура валидации классификации для подтверждения отсутствия неклассифицированных записей (Unknown).*

*Во избежание survivorship bias и acquisition bias snapshots формируются только для клиентов, привлечённых до соответствующей даты наблюдения. Клиенты, совершившие первую покупку после даты snapshot, исключаются из анализа, поскольку на момент наблюдения ещё не являлись частью активной клиентской базы.*

**Snapshot validation**

Для каждого monthly snapshot была проверена корректность распределения клиентов по lifecycle-сегментам.

Во всех snapshot-срезах:
- отсутствуют пользователи с неопределенным статусом (Unknown = 0);
- количество клиентов последовательно растет от месяца к месяцу;
- постепенно формируется churn-сегмент по мере накопления исторических данных;
- распределение между Active / At Risk / Loyal выглядит стабильно и бизнес-реалистично.

Также наблюдается:
- рост общего числа клиентов;
- постепенное увеличение доли клиентов в состоянии Churn;
- умеренный рост сегмента Loyal.

Это подтверждает корректность логики snapshot segmentation и позволяет перейти к анализу миграций между lifecycle stages.

## Customer Lifecycle Transition Matrix Analysis

После формирования последовательных lifecycle snapshots необходимо оценить, насколько устойчивыми являются различные сегменты клиентской базы и каким образом пользователи перемещаются между стадиями жизненного цикла. Именно динамика переходов позволяет понять, где формируются основные механизмы удержания, деградации и восстановления клиентов, которые в дальнейшем определяют структуру churn и долгосрочную динамику выручки.

Цель анализа — выявить наиболее вероятные направления customer migration между lifecycle stages и оценить стабильность каждого состояния жизненного цикла. Дополнительно проверяется гипотеза о том, что зрелые сегменты демонстрируют более высокий уровень удержания, тогда как отдельные стадии выступают промежуточными зонами риска, через которые проходит основная часть будущего оттока.

Для анализа используются transition matrices между последовательными snapshot-периодами. Матрицы отражают вероятность перехода клиентов из одного lifecycle-состояния в другое и позволяют количественно оценить стабильность retention, скорость деградации клиентской базы и интенсивность восстановления пользователей. Помимо отдельных периодов дополнительно рассчитывается агрегированная transition matrix, объединяющая все наблюдения и позволяющая выделить устойчивые поведенческие закономерности независимо от особенностей отдельных временных интервалов.

Ожидается, что сегмент Loyal будет характеризоваться наиболее высокой вероятностью сохранения статуса, тогда как стадии Active и At Risk продемонстрируют повышенную мобильность и станут основными источниками дальнейшей lifecycle-деградации. Подтверждение данной закономерности позволит локализовать ключевые точки потери клиентов и определить сегменты, в которых retention-интервенции потенциально способны обеспечить максимальный бизнес-эффект.

Рассмотрим матрицы переходов между lifecycle stages и оценим устойчивость клиентской базы на различных этапах жизненного цикла.

```python
# =====================================================
# PRINT MATRICES
# =====================================================

for tm in transition_matrices:

    print("\n")
    print("="*60)
    print(tm['period'])
    print("="*60)

    print((tm['matrix'] * 100).round(2))

# =====================================================
# AGGREGATED MATRIX
# =====================================================

all_transitions = pd.concat([
    tm['raw_data']
    for tm in transition_matrices
])

aggregated_matrix = pd.crosstab(
    all_transitions['Lifecycle_Stage_from'],
    all_transitions['Lifecycle_Stage_to'],
    normalize='index'
)

print("\n")
print("="*60)
print("AGGREGATED TRANSITION MATRIX")
print("="*60)

print((aggregated_matrix * 100).round(2))
```

*Нормализация по строкам (normalize='index') интерпретирует каждую строку как условное распределение вероятностей перехода из исходного состояния (from state) в последующее (to state). Такой подход позволяет анализировать retention, churn dynamics и вероятность миграции клиентов между уровнями активности независимо от абсолютного размера сегментов.*

**Key Findings**

Агрегированная матрица переходов показала высокую неоднородность устойчивости различных lifecycle-сегментов. Для сегмента Churn показатель стабильности еще выше — 95.7%, что свидетельствует о крайне низкой вероятности самостоятельного возврата ранее потерянных клиентов.

Наибольшая динамика наблюдается в сегменте Active. Лишь 61.9% клиентов сохраняют данный статус в следующем периоде, тогда как 31.1% переходят в At Risk. Таким образом, вероятность деградации активного клиента более чем в четыре раза превышает вероятность его продвижения к устойчивой лояльности.

Сегмент At Risk выступает ключевой переходной зоной жизненного цикла. Около 23.4% клиентов данного сегмента переходят в Churn уже в следующем периоде, тогда как вероятность восстановления до Active составляет лишь 7.1%. Это делает At Risk основным каналом формирования будущего оттока клиентской базы.

Дополнительно анализ показывает высокую эффективность первичной активации новых клиентов. После первой покупки 88.2% пользователей переходят в Active, однако только 7.3% достигают стадии Loyal уже в следующем lifecycle-периоде.

Полученные результаты указывают на наличие выраженной lifecycle-воронки внутри клиентской базы.

Основной разрыв возникает на этапе Active → Loyal, где значительная часть клиентов не формирует долгосрочную покупательскую привычку и начинает постепенно терять вовлеченность.

Несмотря на продолжающуюся активность клиентов, именно здесь формируется основной поток пользователей, впоследствии переходящих в At Risk. Это означает, что признаки деградации клиентского поведения возникают существенно раньше, чем клиент официально достигает стадии риска.

Такая структура характерна для зрелых потребительских бизнесов, где ключевым драйвером роста становится не реактивация ушедших пользователей, а успешное продвижение клиентов по ранним стадиям жизненного цикла.

Active → At Risk — основной канал формирования будущего churn;

При этом работа с уже ушедшими клиентами обладает ограниченным потенциалом возврата вследствие крайне высокой устойчивости состояния Churn.

## Customer Lifecycle Flow Dynamics

На предыдущем этапе были рассчитаны вероятности переходов между lifecycle-состояниями и выявлены наиболее устойчивые и уязвимые сегменты клиентской базы. Однако матрицы переходов позволяют оценить лишь относительные вероятности миграции пользователей и не показывают масштаб самих потоков клиентов между состояниями.

Цель данного этапа — визуализировать структуру customer lifecycle migration и определить, какие направления движения пользователей формируют основную динамику клиентского портфеля. Ключевой аналитический вопрос заключается в том, через какие lifecycle-переходы происходит удержание, развитие или деградация клиентской базы и где концентрируются наиболее значимые потоки пользователей.

Для анализа используется агрегированный customer flow между последовательными snapshot-периодами. Каждый клиент рассматривается как единица перехода между lifecycle stages, а совокупность всех переходов объединяется в единую карту миграции пользователей. Такой подход позволяет одновременно оценить масштаб retention-потоков, интенсивность деградации клиентов, объемы продвижения пользователей к более ценным стадиям жизненного цикла и устойчивость отдельных сегментов.

Ожидается, что основная часть клиентских потоков будет сосредоточена вокруг сохранения текущего lifecycle-статуса, тогда как наиболее значимые потери будут происходить через последовательную деградацию Active → At Risk → Churn. Если данная структура подтвердится, это будет свидетельствовать о том, что отток является не одномоментным событием, а результатом постепенного ухудшения пользовательской активности, создающего окно возможностей для ранних retention-интервенций.

Для более наглядной оценки структуры клиентских миграций визуализируем агрегированный Customer Lifecycle Flow и проанализируем ключевые направления движения пользователей между lifecycle-сегментами.

```python
# =====================================================
# SANKEY DIAGRAM
# =====================================================

# ---------------------------------------------
# AGGREGATE TRANSITIONS
# ---------------------------------------------

sankey_data = (
    all_transitions
    .groupby([
        'Lifecycle_Stage_from',
        'Lifecycle_Stage_to'
    ])
    .size()
    .reset_index(name='count')
)

# ---------------------------------------------
# STAGE LABELS
# ---------------------------------------------

stages = [
    'New',
    'Active',
    'Loyal',
    'At Risk',
    'Churn'
]

# create index mapping
stage_to_idx = {
    stage: idx
    for idx, stage in enumerate(stages)
}

# ---------------------------------------------
# BUILD LINKS
# ---------------------------------------------

source = sankey_data['Lifecycle_Stage_from'].map(stage_to_idx)

target = sankey_data['Lifecycle_Stage_to'].map(stage_to_idx)

value = sankey_data['count']

# ---------------------------------------------
# SANKEY FIGURE
# ---------------------------------------------

fig = go.Figure(
    data=[
        go.Sankey(
            node=dict(
                pad=20,
                thickness=25,
                line=dict(color='black', width=0.5),
                label=stages
            ),

            link=dict(
                source=source,
                target=target,
                value=value
            )
        )
    ]
)

# ---------------------------------------------
# LAYOUT
# ---------------------------------------------

fig.update_layout(
    title='Customer Lifecycle Transition Flow',
    font_size=12,
    height=700,
    width=1000
)

fig.show()
```

Основной путь деградации клиентской базы проходит через последовательность Active → At Risk → Churn. Эти потоки существенно превышают объемы обратного восстановления клиентов.

Одновременно наблюдается ограниченная, но устойчивая восходящая мобильность. Около 2.2 тыс. новых клиентов переходят в Active, а часть Active-клиентов достигает стадии Loyal. Однако масштабы продвижения вверх по lifecycle заметно уступают объемам деградационных переходов.

**Business Interpretation**

Одновременно высокая стабильность Loyal-сегмента подтверждает, что после формирования устойчивой покупательской привычки вероятность дальнейшего удержания резко возрастает.

**Strategic Implications**

Фактически lifecycle-модель показывает, что долгосрочный рост выручки определяется не масштабом привлечения новых клиентов, а способностью компании замедлять деградационные переходы и ускорять движение пользователей к состоянию устойчивой лояльности.

**Key Takeaway**

Customer lifecycle обладает высокой предсказуемостью: клиенты преимущественно движутся по траектории New → Active → Loyal либо Active → At Risk → Churn. Поэтому основная возможность роста находится в управлении скоростью этих переходов. Контроль миграции между lifecycle stages становится более сильным инструментом управления будущей выручкой, чем мониторинг статической структуры клиентской базы.

# Part 2. Revenue Dynamics Across Lifecycle Transitions 

## Revenue Transition Matrix: Economic Flow Across Lifecycle Stages

На предыдущих этапах были исследованы вероятности миграции клиентов между lifecycle-состояниями и выявлены основные траектории развития и деградации клиентской базы. Однако сами по себе пользовательские переходы не позволяют оценить их экономическую значимость. Для бизнеса принципиально важно понимать не только куда перемещаются клиенты, но и какие денежные потоки сопровождают эти изменения поведения.

Цель данного этапа — определить, какие lifecycle-переходы формируют основную часть текущей выручки, где происходит наибольший прирост клиентской ценности и какие миграции генерируют наиболее существенные потери дохода. Дополнительно проверяется гипотеза о том, что ключевые изменения выручки связаны не столько с привлечением новых клиентов или формальным churn, сколько с изменением покупательской активности внутри существующей клиентской базы.

Для анализа каждая lifecycle-миграция рассматривается одновременно в двух измерениях: пользовательском и финансовом. Помимо объема выручки, проходящей через конкретный переход, дополнительно оценивается изменение клиентской ценности между последовательными snapshot-периодами. Такой подход позволяет связать customer behavior с динамикой revenue и перейти от анализа клиентских потоков к анализу денежных потоков жизненного цикла.

Ожидается, что основная часть выручки будет сосредоточена в переходах, связанных с удержанием наиболее ценных сегментов, прежде всего внутри Active и Loyal. Одновременно существенные отрицательные изменения выручки могут концентрироваться в переходах, отражающих постепенную деградацию клиентской активности, что позволит количественно оценить реальные источники revenue leakage внутри клиентского портфеля.

Рассчитаем матрицу revenue transitions и определим, какие lifecycle-переходы выступают ключевыми драйверами роста и потерь выручки компании.

```python
# =====================================================
# BUILD REVENUE TRANSITIONS
# =====================================================

transition_dfs = []

for i in range(len(snapshot_dates)-1):

    current_date = snapshot_dates[i]
    next_date = snapshot_dates[i+1]

    current_snapshot = all_snapshots[
        all_snapshots['Snapshot_Date'] == current_date
    ][[
        'card',
        'Lifecycle_Stage',
        'Monetary'
    ]]

    next_snapshot = all_snapshots[
        all_snapshots['Snapshot_Date'] == next_date
    ][[
        'card',
        'Lifecycle_Stage',
        'Monetary'
    ]]

    merged = current_snapshot.merge(
        next_snapshot,
        on='card',
        how='inner',
        suffixes=('_from', '_to')
    )

    merged['Revenue_Change'] = (
        merged['Monetary_to']
        -
        merged['Monetary_from']
    )

    merged['Period'] = (
        f"{current_date.date()}"
        +
        " -> "
        +
        f"{next_date.date()}"
    )

    transition_dfs.append(merged)

all_transitions = pd.concat(
    transition_dfs
)

# =====================================================
# REVENUE TRANSITION MATRIX
# =====================================================

revenue_matrix = (
    all_transitions
    .groupby([
        'Lifecycle_Stage_from',
        'Lifecycle_Stage_to'
    ])
    .agg({
        'Monetary_to': 'sum',
        'Revenue_Change': 'sum',
        'card': 'count'
    })
    .reset_index()
)

revenue_matrix.columns = [
    'From',
    'To',
    'Revenue',
    'Revenue_Change',
    'Customers'
]

#print("\n")
print("="*60)
print("REVENUE TRANSITION MATRIX ANALYSIS")
print("="*60)

print(
    revenue_matrix
    .sort_values(
        'Revenue',
        ascending=False
    )
)
```

**Key Findings**

Экономика клиентской базы концентрируется вокруг удержания зрелых клиентов. Переход Loyal → Loyal генерирует 27.5 млн руб. выручки — более чем в 3 раза больше любого другого lifecycle-потока. Одновременно этот сегмент продолжает наращивать ценность (+0.98 млн руб.), подтверждая высокий уровень monetization retention.

Основной источник роста выручки формируется не за счет привлечения новых клиентов, а за счет развития существующей базы. Наибольший положительный вклад создают переходы Active → Loyal (+0.67 млн руб.), New → Loyal (+0.48 млн руб.) и Active → Active (+0.32 млн руб.), что указывает на значимую роль expansion внутри удержанных клиентов.

Крупнейшие потери сосредоточены в деградационных переходах. Максимальный revenue leakage возникает при переходе At Risk → Churn (-1.51 млн руб.), а также в потоках At Risk → At Risk (-0.88 млн руб.), Loyal → Active (-0.34 млн руб.) и Loyal → At Risk (-0.30 млн руб.). Потери начинают формироваться задолго до фактического churn.

**Business Interpretation**

Результаты показывают, что основная часть экономической ценности создается через удержание и развитие существующих клиентов, а не через acquisition. Рост выручки возникает преимущественно тогда, когда клиенты продвигаются к более зрелым стадиям жизненного цикла или сохраняют высокий уровень вовлеченности.

При этом наиболее опасным процессом является не сам churn, а предшествующая ему деградация поведения. Значительная часть выручки теряется еще на этапах снижения частоты покупок и перехода клиентов в состояние риска. Формальный churn выступает финальной стадией уже реализованной потери клиентской ценности.

**Strategic Implications**

Для роста бизнеса критично управлять не количеством клиентов, а направлением lifecycle-миграций.

Приоритетными зонами воздействия являются:  
- Ускорение переходов Active → Loyal как крупнейшего источника органического роста revenue.  
- Снижение потоков Active → At Risk и Loyal → At Risk как ранних источников снижения revenue.  
- Превентивная работа с сегментом At Risk до перехода в Churn, где концентрируется максимальный объем прямых потерь.  

**Executive Takeaway**

Revenue decomposition показывает, что экономика сети является прежде всего retention-driven, а не acquisition-driven. Основной рост выручки создают удержанные клиенты, увеличивающие свою ценность внутри жизненного цикла, тогда как основные потери возникают через постепенную деградацию активной базы. Следовательно, ключевой рычаг роста находится в управлении lifecycle-переходами и предотвращении раннего снижения клиентской активности, а не в компенсации потерь новым привлечением.

## Revenue Growth & Revenue Loss Drivers

На предыдущем этапе были определены денежные потоки между lifecycle-состояниями и рассчитан вклад каждого transition в изменение выручки. Однако агрегированная матрица переходов затрудняет выделение наиболее значимых факторов роста и потерь бизнеса.

Цель данного этапа — определить, какие именно изменения пользовательского поведения оказывают максимальное влияние на динамику revenue. Проверим, какие lifecycle transitions формируют основной прирост выручки, а какие являются ключевыми источниками её снижения. Особый интерес представляет сравнение вклада retention, customer expansion, degradation и churn, поскольку именно эти механики определяют эффективность growth и retention-стратегии компании.

Для анализа рассмотрим абсолютное изменение выручки (Revenue Change) по каждому типу перехода между lifecycle-состояниями. ## Revenue Driver Analysis (Flow Classification Model)

На предыдущих этапах были выявлены ключевые пользовательские переходы между lifecycle-состояниями и оценен их вклад в изменение выручки. Однако отдельные transition-сценарии затрудняют стратегическую интерпретацию результатов, поскольку десятки различных переходов необходимо свести к ограниченному набору бизнес-драйверов, которыми компания может управлять через CRM, retention и marketing initiatives.

Цель данного этапа — определить, какие механизмы клиентской динамики формируют рост и снижение выручки на уровне бизнеса. Иными словами, необходимо ответить на вопрос: за счет каких behavioral drivers компания зарабатывает и теряет деньги — привлечения новых клиентов, удержания существующей базы, реактивации неактивных пользователей, расширения потребления или оттока клиентов.

Для этого каждый lifecycle transition классифицируется в одну из укрупненных категорий revenue drivers: Acquisition, Reactivation, Expansion, Contraction, Churn или Stable. Такой подход позволяет перейти от анализа отдельных пользовательских миграций к оценке фундаментальных источников revenue growth и revenue loss, сопоставимых с классическими growth-моделями бизнеса.

Ожидается, что основная часть положительной динамики будет формироваться за счет retention и expansion существующей клиентской базы, тогда как наиболее существенные потери будут связаны с churn и снижением покупательской активности retained customers. Результаты анализа позволят количественно оценить вклад каждого драйвера в изменение выручки и определить приоритетные направления для retention-driven growth стратегии.

Блок отвечает на вопрос: «Какие типы клиентских переходов создают, расширяют или уничтожают revenue?»

Сформируем классификацию lifecycle transitions по ключевым revenue drivers и оценим их совокупный вклад в динамику бизнеса.

```python
# =====================================================
# FLOW CLASSIFICATION
# =====================================================

def classify_flow(row):

    from_stage = row['From']
    to_stage = row['To']

    # =========================================
    # ACQUISITION
    # =========================================

    if to_stage == 'New':
        return 'Acquisition'

    # =========================================
    # REACTIVATION
    # =========================================

    if (
        from_stage in ['At Risk', 'Churn']
        and
        to_stage in ['Active', 'Loyal']
    ):
        return 'Reactivation'

    # =========================================
    # CHURN
    # =========================================

    if to_stage == 'Churn':
        return 'Churn'

    # =========================================
    # EXPANSION
    # =========================================

    if row['Revenue_Change'] > 0:
        return 'Expansion'

    # =========================================
    # CONTRACTION
    # =========================================

    if row['Revenue_Change'] < 0:
        return 'Contraction'

    # =========================================
    # STABLE
    # =========================================

    return 'Stable'

# =====================================================
# APPLY CLASSIFICATION
# =====================================================

revenue_matrix['Revenue_Driver'] = (

    revenue_matrix.apply(
        classify_flow,
        axis=1
    )

)

# =====================================================
# DRIVER SUMMARY
# =====================================================

driver_summary = (

    revenue_matrix
    .groupby(
        'Revenue_Driver',
        observed=True
    )
    .agg({
        'Revenue_Change': 'sum',
        'Customers': 'sum',
        'Revenue': 'sum'
    })
    .reset_index()

)

# =====================================================
# SORT
# =====================================================

driver_summary = driver_summary.sort_values(
    'Revenue_Change',
    ascending=False
)

# =====================================================
# RENAME COLUMNS
# =====================================================

driver_summary.columns = [
    'Revenue_Driver',
    'Total_Revenue_Change',
    'Total_Customers',
    'Total_Revenue'
]

# =====================================================
# ROUND
# =====================================================

driver_summary = driver_summary.round(2)

# =====================================================
# OUTPUT
# =====================================================

#print("\n")
print("="*60)
print("REVENUE DRIVER ANALYSIS")
print("="*60)

print(driver_summary)
```

**Key Findings**

Динамика выручки определяется преимущественно качеством работы с существующей клиентской базой, а не привлечением новых клиентов. Основным источником роста выступают retention и expansion текущих покупателей, тогда как основные потери связаны с постепенной деградацией клиентской активности и последующим churn.

**Growth Drivers**

Expansion является крупнейшим драйвером роста выручки (+2.74 млн), формируя более 85% всего положительного эффекта. Рост обеспечивается преимущественно удержанными клиентами, которые увеличивают частоту покупок или объем потребления.

Reactivation вносит дополнительный положительный вклад (+355 тыс.), подтверждая наличие экономического потенциала у клиентов, ранее находившихся в состоянии деградации или оттока.

Acquisition оказывает минимальное влияние на итоговую динамику (+48 тыс.), что свидетельствует о существенно более высокой значимости развития существующей базы по сравнению с привлечением новых покупателей.

**Revenue Loss Drivers**

Крупнейшим источником потерь является Contraction (-1.69 млн), превышающий прямые потери от churn. Это означает, что снижение выручки начинается задолго до фактической потери клиента и проявляется через постепенное сокращение покупательской активности retained клиентов.

Churn генерирует дополнительные потери в размере -1.51 млн и является скорее финальной стадией уже начавшейся деградации клиентского поведения.

**Business Interpretation**

Анализ показывает, что ключевая проблема бизнеса заключается не в недостаточном привлечении новых клиентов, а в управлении жизненным циклом существующей базы.

На каждый рубль потерь от прямого churn приходится более одного рубля потерь вследствие contraction, что смещает фокус retention-стратегии с реактивной борьбы с оттоком на более раннее выявление признаков снижения активности.

**Strategic Implication**

Наибольший потенциальный эффект для бизнеса способны обеспечить:  
1. предотвращение переходов Active → At Risk и Loyal → At Risk;  
2. раннее выявление contraction-сценариев до наступления churn;  
3. развитие программ expansion для удержанных клиентов;  
4. системная реактивация сегментов At Risk и Churn.  

Результаты подтверждают, что рост выручки в рассматриваемой модели носит преимущественно retention-driven характер, а основная экономическая ценность сосредоточена внутри существующей клиентской базы.

## Retained Customer Economics (Stable Lifecycle Monetization Analysis)

На предыдущем этапе была выполнена декомпозиция изменения выручки по ключевым business drivers и определен вклад acquisition, reactivation, expansion, contraction и churn в общую динамику бизнеса. Однако такой подход не позволяет отделить эффект удержания клиентов от эффекта изменения их покупательского поведения после удержания.

Цель данного этапа — оценить экономику удержанной клиентской базы и определить, какой вклад в изменение выручки вносят клиенты, сохранившие свое активное lifecycle-состояние. Основной аналитический вопрос заключается в том, создает ли retention дополнительную ценность через рост потребления или удержанные пользователи со временем сокращают объем покупок.

Для анализа рассматриваются только клиенты, сохранившие статус Active или Loyal между соседними snapshot-периодами. Внутри этой группы изменение выручки разделяется на два типа поведения: Expansion — рост расходов удержанных клиентов и Contraction — снижение расходов при сохранении активности. Такой подход позволяет изолировать чистый monetization effect от влияния acquisition, churn и lifecycle migration.

Ожидается, что наиболее зрелые и лояльные клиенты будут формировать основную часть expansion-эффекта. Если же contraction окажется сопоставимым с expansion, это будет указывать на скрытую деградацию покупательского поведения даже при формально высоком retention.

Блок отвечает на вопрос: «Как изменяется выручка внутри стабильной удержанной клиентской базы без влияния lifecycle-переходов?»

Оценим вклад expansion и contraction внутри удержанной клиентской базы и проанализируем экономическую эффективность retention-модели.

```python
# =====================================================
# TRUE RETAINED CUSTOMERS
# =====================================================

retained = all_transitions[

    (
        all_transitions['Lifecycle_Stage_from']
        ==
        all_transitions['Lifecycle_Stage_to']
    )

    &

    (
        all_transitions['Lifecycle_Stage_from']
        .isin(['Active', 'Loyal'])
    )

].copy()

# =====================================================
# REVENUE CHANGE
# =====================================================

retained['Revenue_Change'] = (

    retained['Monetary_to']
    -
    retained['Monetary_from']

)

# =====================================================
# EXPANSION / CONTRACTION
# =====================================================

retained['Revenue_Type'] = np.where(

    retained['Revenue_Change'] > 0,
    'Expansion',
    'Contraction'

)

# =====================================================
# SUMMARY
# =====================================================

expansion_summary = (

    retained
    .groupby(
        'Revenue_Type',
        observed=True
    )['Revenue_Change']
    .sum()

)

#print("\n")
print("="*60)
print("RETAINED CUSTOMER ECONOMICS")
print("="*60)

print(expansion_summary)
```

### Expansion vs Contraction within Retained Customers

На предыдущем этапе была рассчитана чистая экономика удержанной клиентской базы и определен вклад expansion и contraction в изменение выручки среди клиентов, сохранивших активный lifecycle-статус.

Для упрощения интерпретации результатов визуализируем баланс между положительным и отрицательным вкладом retained customers. Такой взгляд позволяет быстро оценить, создает ли удержанная база дополнительную выручку или значительная часть retention-эффекта компенсируется снижением покупательской активности.

Визуализируем соотношение expansion и contraction внутри удержанных клиентов.

```python
# =====================================================
# PREPARE VALUES
# =====================================================

expansion_total = expansion_summary.get('Expansion', 0)

contraction_total = abs(
    expansion_summary.get('Contraction', 0)
)

# =====================================================
# STACKED BAR CHART
# =====================================================

fig = go.Figure()

# Expansion
fig.add_trace(

    go.Bar(

        name='Expansion',

        x=['Retained Customers'],

        y=[expansion_total],
 
        marker_color='#009E73'   #2ca02c

    )

)

# Contraction
fig.add_trace(

    go.Bar(

        name='Contraction',

        x=['Retained Customers'],

        y=[-contraction_total],

        marker_color='#F28E2B' 

    )

)

# =====================================================
# LAYOUT
# =====================================================

fig.update_layout(

    title='Expansion vs Contraction Revenue',

    barmode='relative',

    yaxis_title='Revenue Change',

    template='plotly_white',

    height=500,

    width=1000

)

fig.show()
```

**Key Findings**

Удержанная клиентская база является нетто-источником роста выручки. Несмотря на наличие существенного contraction-эффекта, увеличение расходов части retained customers более чем компенсирует снижение потребления остальных клиентов.

**Expansion vs Contraction Balance**

Внутри удержанной базы expansion сформировал +3.33 млн выручки, тогда как contraction привел к потерям в размере -2.03 млн. Чистый эффект составил около +1.30 млн.

Это означает, что retention в рассматриваемой модели не ограничивается сохранением клиентской активности, а сопровождается дополнительной монетизацией значительной части удержанных пользователей.

**Business Interpretation**

Даже среди клиентов, сохранивших статус Active или Loyal, наблюдается выраженная неоднородность покупательского поведения. Удержание само по себе не гарантирует сохранения выручки: часть retained customers наращивает потребление, тогда как другая часть постепенно сокращает объем покупок.

Однако суммарный положительный эффект expansion существенно превышает потери от contraction, что свидетельствует о высокой экономической устойчивости активной клиентской базы.

**Strategic Implication**

Основной потенциал роста сосредоточен внутри существующих клиентов. Приоритетом retention-стратегии должно быть не только предотвращение churn, но и системное развитие expansion-механик для Active и Loyal сегментов, поскольку именно они формируют значительную часть органического роста выручки.

## Revenue Growth Decomposition (Accounting Decomposition Model)

На предыдущих этапах были определены основные драйверы изменения выручки на уровне lifecycle-переходов и отдельно исследована экономика удержанных клиентов. Теперь перейдем к более высокоуровневому представлению, которое позволяет разложить совокупное изменение выручки на ключевые компоненты роста бизнеса.

Цель анализа — определить, какая часть выручки формируется за счет привлечения новых клиентов, возврата ранее потерянной аудитории, расширения потребления существующей клиентской базы, а также какие потери возникают вследствие сокращения расходов клиентов и их полного оттока. Фактически анализ отвечает на вопрос: какие механизмы создают рост выручки и какие механизмы его ограничивают.

Для этого каждая клиентская динамика будет отнесена к одному из пяти компонентов: New Revenue, Reactivated Revenue, Expansion Revenue, Contraction Revenue и Churned Revenue. Такой подход соответствует логике decomposition-моделей, широко используемых в subscription-, SaaS- и retention-driven бизнесах для оценки качества роста.

Полученная структура позволит отделить рост, обеспеченный привлечением клиентов, от роста внутри существующей базы, а также определить, какие источники потерь оказывают наибольшее давление на выручку.

Блок отвечает на вопрос: «Из каких фундаментальных growth-компонентов складывается экономика бизнеса?»

Рассчитаем вклад каждого компонента в общую экономику клиентской базы.

```python
# =====================================================
# Revenue Change
# =====================================================

all_transitions['Revenue_Change'] = (
    all_transitions['Monetary_to']
    -
    all_transitions['Monetary_from']
)

# =====================================================
# MASKS
# =====================================================

# =====================================================
# NEW
# =====================================================

new_mask = (
    all_transitions['Lifecycle_Stage_to'] == 'New'
)

# =====================================================
# REACTIVATION
# =====================================================

reactivation_mask = (
    (
        all_transitions['Lifecycle_Stage_from']
        .isin(['At Risk', 'Churn'])
    )
    &
    (
        all_transitions['Lifecycle_Stage_to']
        .isin(['Active', 'Loyal'])
    )
)

# =====================================================
# CHURN
# =====================================================

churn_mask = (
    all_transitions['Lifecycle_Stage_to'] == 'Churn'
)

# =====================================================
# RETAINED BASE
# =====================================================

retained_mask = (
    (~new_mask)
    &
    (~reactivation_mask)
    &
    (~churn_mask)
)

# =====================================================
# REVENUE ATTRIBUTION
# =====================================================

new_revenue = all_transitions[
    new_mask
]['Monetary_to'].sum()

reactivated_revenue = all_transitions[
    reactivation_mask
]['Monetary_to'].sum()

churned_revenue = all_transitions[
    churn_mask
]['Monetary_from'].sum()

retained = all_transitions[
    retained_mask
].copy()

# =====================================================
# EXPANSION REVENUE
# =====================================================

expansion_revenue = retained[
    retained['Revenue_Change'] > 0
]['Revenue_Change'].sum()

# =====================================================
# CONTRACTION REVENUE
# =====================================================

contraction_revenue = abs(

    retained[
        retained['Revenue_Change'] < 0
    ]['Revenue_Change'].sum()

)
# =====================================================
# Final Table
# =====================================================

decomposition = pd.DataFrame({

    'Component': [
        'New Revenue',
        'Reactivated Revenue',
        'Expansion Revenue',
        'Contraction Revenue',
        'Churned Revenue'
    ],

    'Value': [
        new_revenue,
        reactivated_revenue,
        expansion_revenue,
        contraction_revenue,
        churned_revenue
    ]
})
#print("\n")
print("="*60)
print("REVENUE DECOMPOSITION ANALYSIS")
print("="*60)

print(decomposition)
```

### Revenue Growth Decomposition

На предыдущем этапе выручка была декомпозирована на ключевые growth-компоненты: привлечение новых клиентов, реактивацию, расширение потребления существующей базы, сокращение расходов клиентов и потери вследствие оттока.

Для упрощения интерпретации визуализируем полученную структуру в формате Waterfall Chart. Такой формат позволяет одновременно оценить относительный вклад каждого драйвера и увидеть баланс между источниками роста и факторами снижения выручки.

Особый интерес представляет соотношение Expansion Revenue и Contraction/Churned Revenue, поскольку именно оно отражает способность клиентской базы генерировать органический рост и компенсировать естественные потери.

Построим waterfall-диаграмму структуры изменения выручки.

```python
# ============================================
# COMPONENTS
# ============================================

components = {
    'New Revenue': new_revenue,
    'Reactivated Revenue': reactivated_revenue,
    'Expansion Revenue': expansion_revenue,
    'Contraction Revenue': -contraction_revenue,
    'Churned Revenue': -churned_revenue
}

labels = list(components.keys())
values = list(components.values())

# ============================================
# WATERFALL
# ============================================

fig = go.Figure(go.Waterfall(

    name='Revenue',

    orientation='v',

    measure=[
        'relative',
        'relative',
        'relative',
        'relative',
        'relative'
    ],

    x=labels,

    y=values,

    text=[
        f"{v:,.0f}"
        for v in values
    ],

    textposition='outside'
))

fig.update_layout(

    title='Revenue Decomposition',

    waterfallgap=0.3,

    height=700,
    width=1100
)

fig.show()
```

**Key Findings**

Основным источником роста выручки является существующая клиентская база. Expansion Revenue (+4.79 млн) обеспечивает более 77% всего положительного вклада, что указывает на высокий потенциал развития текущих клиентов без критической зависимости от привлечения новой аудитории.
Reactivated Revenue (+1.26 млн) существенно превышает New Revenue (+0.18 млн). Возврат ранее потерянных клиентов генерирует значительно больший экономический эффект, чем первичное привлечение, что делает реактивацию одним из наиболее эффективных growth-направлений.
Ключевой источник потерь — снижение расходов внутри действующей базы. Contraction Revenue (-3.74 млн) более чем в 2 раза превышает потери от полного оттока (Churned Revenue -1.51 млн).

**Business Interpretation**

Проблема бизнеса сосредоточена не столько в потере клиентов, сколько в постепенном снижении их покупательской активности. Значительная часть выручки теряется до момента фактического перехода клиента в churn.

Одновременно клиентская база демонстрирует выраженный потенциал монетизации: рост среднего чека и реактивация клиентов формируют основной положительный вклад в выручку.

**Strategic Implications**

Приоритетом должен стать контроль ранних признаков снижения revenue в сегментах Active и Loyal до перехода клиентов в At Risk.
Реактивационные механики обладают высокой экономической отдачей и могут приносить больший эффект, чем дополнительное масштабирование acquisition.
Для ускорения роста наибольший потенциал находится в программах expansion: персонализированных предложениях, увеличении частоты покупок и развитии высокоценных клиентов.

**Key Takeaway**

Рост выручки формируется преимущественно за счет монетизации и возврата существующей клиентской базы, тогда как главный риск связан с сокращением расходов текущих клиентов, а не с их прямым оттоком. Это характерно для зрелой retention-driven модели бизнеса, где эффективность работы с базой важнее объема нового привлечения.

## Revenue Movement Across Customer Lifecycle States

На предыдущих этапах были определены основные драйверы изменения выручки и оценен вклад каждого компонента в общий финансовый результат. Однако оба анализа отвечают на вопрос «что изменило выручку», но не показывают «где именно внутри клиентского жизненного цикла происходят эти изменения».

На данном этапе исследуем структуру денежных потоков между lifecycle-сегментами. Цель анализа — определить, через какие переходы клиентов проходит основная часть выручки, какие сегменты выступают ключевыми источниками и получателями ценности, а также где формируются крупнейшие точки роста и потерь.

Для этого агрегируем все межсегментные переходы и рассматриваем их одновременно в двух измерениях: объеме выручки, проходящей через каждый поток, и финансовом эффекте соответствующего перехода. Такой подход позволяет перейти от анализа отдельных драйверов к анализу всей lifecycle-экономики как единой системы взаимосвязанных потоков.

Ожидается, что визуализация позволит выявить критические маршруты движения клиентской базы, определить наиболее ценные траектории развития клиентов и локализовать сегменты, в которых сосредоточены основные риски деградации выручки.

Построим карту денежных потоков между lifecycle-сегментами и оценим их вклад в общую структуру монетизации клиентской базы.

```python
# =====================================================
# REVENUE SANKEY
# =====================================================
# ---------------------------------------------
# AGGREGATE REVENUE FLOWS
# ---------------------------------------------

revenue_flows = (
    all_transitions
    .groupby([
        'Lifecycle_Stage_from',
        'Lifecycle_Stage_to'
    ])
    .agg({
        'Revenue_Change': 'sum',
        'Monetary_to': 'sum'
    })
    .reset_index()
)

# ---------------------------------------------
# STAGES
# ---------------------------------------------

stages = [
    'New',
    'Active',
    'Loyal',
    'At Risk',
    'Churn'
]

stage_to_idx = {
    stage: idx
    for idx, stage in enumerate(stages)
}

# ---------------------------------------------
# BUILD LINKS
# ---------------------------------------------

source = revenue_flows[
    'Lifecycle_Stage_from'
].map(stage_to_idx)

target = revenue_flows[
    'Lifecycle_Stage_to'
].map(stage_to_idx)

# thickness of flow
value = revenue_flows[
    'Monetary_to'
].abs()

# ---------------------------------------------
# LINK COLORS
# ---------------------------------------------

link_colors = np.where(

    revenue_flows['Revenue_Change'] >= 0,

    'rgba(0, 158, 115, 1.0)',   # green    
 
    'rgba(244, 164, 96, 1.0)'    # red   
)

# ---------------------------------------------
# OPTIONAL: HOVER LABELS
# ---------------------------------------------

hover_text = (
    revenue_flows['Lifecycle_Stage_from']
    + ' → '
    + revenue_flows['Lifecycle_Stage_to']
    + '<br>Revenue: '
    + revenue_flows['Monetary_to'].round(2).astype(str)
    + '<br>Revenue Change: '
    + revenue_flows['Revenue_Change'].round(2).astype(str)
)

# ---------------------------------------------
# SANKEY FIGURE
# ---------------------------------------------

fig = go.Figure(
    data=[
        go.Sankey(

            node=dict(
                pad=20,
                thickness=25,
                line=dict(
                    color='black',
                    width=0.5
                ),
                label=stages,
                color='lightblue'
            ),

            link=dict(
                source=source,
                target=target,
                value=value,
                color=link_colors,
                customdata=hover_text,
                hovertemplate='%{customdata}<extra></extra>'
            )
        )
    ]
)

# ---------------------------------------------
# LAYOUT
# ---------------------------------------------

fig.update_layout(
    title='Revenue Flow Between Lifecycle Stages',
    font_size=12,
    height=700,
    width=1100
)

fig.show()
```

**Key Findings**

**1. Выручка концентрируется в удержанном лояльном сегменте**  
Переход Loyal → Loyal формирует крупнейший денежный поток системы — 27.5 млн ₽ выручки и +978 тыс. ₽ прироста. Основная часть экономики бизнеса создается не за счет привлечения новых клиентов, а за счет сохранения и развития существующей лояльной базы.

**2. Наиболее ценный путь развития клиента — переход в Loyal**  
Переходы Active → Loyal (+669 тыс. ₽) и New → Loyal (+480 тыс. ₽) обеспечивают один из крупнейших вкладов в рост выручки. Повышение доли клиентов, достигающих стадии Loyal, оказывает существенно больший эффект, чем расширение верхней части воронки.

**3. Основная зона потерь сосредоточена в сегменте At Risk**  
Сегмент At Risk выступает крупнейшим источником деградации выручки:  
- At Risk → Churn: −1.51 млн ₽  
- At Risk → At Risk: −884 тыс. ₽

Потери возникают не только при окончательном оттоке, но и во время длительного пребывания клиентов в состоянии риска.

**4. Реактивация клиентов работает, но имеет ограниченный масштаб**  
Переходы из проблемных сегментов обратно в активные состояния демонстрируют положительный финансовый эффект:  
- At Risk → Active: +144 тыс. ₽  
- At Risk → Loyal: +92 тыс. ₽  
- Churn → Active: +116 тыс. ₽

Однако суммарный объем этих потоков существенно уступает потерям, возникающим внутри сегмента At Risk.  

**Business Interpretation**

Экономика клиентской базы носит выраженный retention-driven характер. Ключевой источник роста — удержание и развитие действующих клиентов, прежде всего внутри сегмента Loyal.

Основной риск для выручки возникает не в момент churn, а на более раннем этапе жизненного цикла, когда клиенты переходят в состояние At Risk и начинают сокращать объем покупок. Сегмент At Risk выступает главным предиктором будущих финансовых потерь.

**Strategic Implications**  

Приоритет №1 — снижение объема клиентов в At Risk через ранние retention-механики.  
Основной growth-рычаг — увеличение конверсии Active → Loyal и New → Loyal.  
Реактивационные кампании имеют положительный ROI, но не способны компенсировать потери от ухудшения состояния действующей базы.  
Для управления выручкой целесообразно использовать долю клиентов в сегментах Loyal и At Risk как опережающие индикаторы будущей динамики дохода.

При этом финансовая динамика определяется не размером клиентской базы как таковой, а распределением клиентов между lifecycle-состояниями. Один дополнительный клиент в сегменте Loyal создает существенно больший долгосрочный эффект, чем один новый привлеченный клиент, тогда как накопление пользователей в At Risk выступает ранним индикатором будущего снижения выручки.

Полученные результаты также показывают, что текущая модель роста является преимущественно органической. Основной прирост revenue обеспечивается за счет повышения ценности существующих клиентов, а не за счет расширения верхней части воронки.

### Влияние на бизнес

Риски  
- Рост доли At Risk-клиентов будет приводить к ускоренному накоплению revenue-at-risk.  
- Сохранение текущих темпов contraction способно ограничивать рост даже при стабильном притоке новых клиентов.  
- Высокая инерционность сегмента Churn снижает эффективность поздних retention-интервенций.  
- Деградация Loyal-клиентов создает непропорционально высокий финансовый ущерб вследствие концентрации выручки в данном сегменте.  

Возможности  
- Ускорение переходов Active → Loyal способно обеспечить наибольший долгосрочный прирост LTV.  
- Снижение потоков Active → At Risk позволит одновременно повысить retention и сократить будущий churn.  
- Масштабирование expansion-механик внутри удержанной базы обладает высоким потенциалом органического роста.  
- Развитие системной реактивации позволяет монетизировать существующую клиентскую базу без дополнительных acquisition-затрат.  

### Ограничения анализа  

- Анализ основан на observational data и не позволяет делать окончательные выводы о причинно-следственных связях между lifecycle-переходами и изменением выручки.
- Результаты чувствительны к выбранным правилам lifecycle-сегментации и используемым порогам RFM-классификации.
- Исследование охватывает один год наблюдений и не учитывает возможные долгосрочные сезонные эффекты.
- Анализ не включает влияние маркетинговых кампаний, ценовых изменений и внешних факторов спроса на наблюдаемую динамику клиентов.

### Приоритетные направления работы  

**1. Построить систему раннего обнаружения деградации клиентов**  
Сформировать predictive retention-модель для выявления клиентов с повышенной вероятностью перехода Active → At Risk и Loyal → At Risk.

**2. Сделать Active-сегмент ключевым объектом CRM-управления**  
Сместить фокус retention-стратегии с реактивной работы с churn на предотвращение ранней деградации поведения.

**3. Развивать механики customer maturation**  
Запустить инициативы, направленные на увеличение вероятности перехода Active → Loyal через персонализацию, программы лояльности и стимулирование регулярных покупок.

**4. Отдельно управлять revenue contraction**  
Включить в систему мониторинга показатели снижения частоты покупок, среднего чека и customer spend как опережающие метрики будущего сокращения revenue.

**5. Масштабировать reactivation-программы**  
Выделить сегменты бывших клиентов с максимальным потенциалом возврата и протестировать автоматизированные reactivation-кампании.

**6. Внедрить lifecycle-driven KPI framework**  
Использовать долю клиентов в сегментах Loyal и At Risk, а также объем revenue-at-risk в качестве ключевых опережающих индикаторов будущей динамики выручки и эффективности retention-стратегии.

### Итоговый вывод

Анализ подтверждает, что ключевой фактор роста аптечной сети — не объем привлечения новых клиентов, а способность компании управлять переходами внутри жизненного цикла существующей базы. Экономическая ценность концентрируется в удержании и развитии Loyal-клиентов, тогда как основные потери возникают через постепенную деградацию активности задолго до формального churn. Следовательно, наиболее эффективная стратегия роста заключается в проактивном управлении lifecycle transitions, развитии клиентской лояльности и раннем предотвращении revenue erosion внутри действующей базы.
